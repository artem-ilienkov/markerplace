```python
# 1. –°—Ñ–æ—Ä–º–∏—Ä—É–µ–º –æ—Å–Ω–æ–≤–Ω–æ–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
engine = create_engine('mysql+mysqlconnector://come_on_dude')

# 1.1 SQL-–∑–∞–ø—Ä–æc –Ω–∞ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è –Ω–µ–¥–µ–ª–∏
query = "SELECT id AS event_id, \
       DATE(createdAt) AS created_date, \
       status, \
       format, \
       type, \
       DATE(startDate) AS start_date, \
       DATE(endDate) AS end_date, \
       name, \
       CASE \
            WHEN INSTR(LOWER(name), '–ú–æ—Å–∫–≤–∞') > 0 OR INSTR(LOWER(location), '–ú–æ—Å–∫–≤–∞') > 0 THEN '–ú–æ—Å–∫–≤–∞' \
            WHEN INSTR(LOWER(name), '–¢–æ–º—Å–∫') > 0 OR INSTR(LOWER(location), '–¢–æ–º—Å–∫') > 0 THEN '–¢–æ–º—Å–∫' \
            WHEN INSTR(LOWER(name), '–ú—Å–∫') > 0 OR INSTR(LOWER(location), '–ú—Å–∫') > 0 THEN '–ú–æ—Å–∫–≤–∞' \
            WHEN INSTR(LOWER(name), '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥') > 0 OR INSTR(LOWER(location), '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥') > 0 THEN '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥' \
            WHEN INSTR(LOWER(name), '–°–ø–±') > 0 OR INSTR(LOWER(location), '–°–ø–±') > 0 THEN '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥' \
            WHEN INSTR(LOWER(name), '–°–∞–Ω–∫—Ç - –ü–µ—Ç–µ—Ä–±—É—Ä–≥') > 0 OR INSTR(LOWER(location), '–°–∞–Ω–∫—Ç - –ü–µ—Ç–µ—Ä–±—É—Ä–≥') > 0 THEN '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥' \
            WHEN INSTR(LOWER(name), '–°–∞–Ω–∫—Ç - –ü–µ—Ç–µ—Ä–±—É—Ä–≥') > 0 OR INSTR(LOWER(location), '–°–∞–Ω–∫—Ç - –ü–µ—Ç–µ—Ä–±—É—Ä–≥') > 0 THEN '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥' \
            WHEN INSTR(LOWER(name), '–°–∞–Ω–∫—Ç- –ü–µ—Ç–µ—Ä–±—É—Ä–≥') > 0 OR INSTR(LOWER(location), '–°–∞–Ω–∫—Ç- –ü–µ—Ç–µ—Ä–±—É—Ä–≥') > 0 THEN '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥' \
            WHEN INSTR(LOWER(name), '–°–∞–Ω–∫—Ç') > 0 OR INSTR(LOWER(location), '–°–∞–Ω–∫—Ç') > 0 THEN '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥' \
            WHEN INSTR(LOWER(name), '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫') > 0 OR INSTR(LOWER(location), '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫') > 0 THEN '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫' \
            WHEN INSTR(LOWER(name), '–ù–æ–≤–æ—Å–∏–±') > 0 OR INSTR(LOWER(location), '–ù–æ–≤–æ—Å–∏–±') > 0 THEN '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫' \
            WHEN INSTR(LOWER(name), '–ù—Å–∫') > 0 OR INSTR(LOWER(location), '–ù—Å–∫') > 0 THEN '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫' \
            WHEN INSTR(LOWER(name), '–ï–∫–∞—Ç–µ—Ä–∏–Ω–±—É—Ä–≥') > 0 OR INSTR(LOWER(location), '–ï–∫–∞—Ç–µ—Ä–∏–Ω–±—É—Ä–≥') > 0 THEN '–ï–∫–∞—Ç–µ—Ä–∏–Ω–±—É—Ä–≥' \
            WHEN INSTR(LOWER(name), '–ù–∏–∂–Ω–∏–π –ù–æ–≤–≥–æ—Ä–æ–¥') > 0 OR INSTR(LOWER(location), '–ù–∏–∂–Ω–∏–π –ù–æ–≤–≥–æ—Ä–æ–¥') > 0 THEN '–ù–∏–∂–Ω–∏–π –ù–æ–≤–≥–æ—Ä–æ–¥' \
            WHEN INSTR(LOWER(name), '–ù–∏–∂–Ω–∏–π-–ù–æ–≤–≥–æ—Ä–æ–¥') > 0 OR INSTR(LOWER(location), '–ù–∏–∂–Ω–∏–π-–ù–æ–≤–≥–æ—Ä–æ–¥') > 0 THEN '–ù–∏–∂–Ω–∏–π –ù–æ–≤–≥–æ—Ä–æ–¥' \
            WHEN INSTR(LOWER(name), '–ù–∏–∂–Ω–∏–π - –ù–æ–≤–≥–æ—Ä–æ–¥') > 0 OR INSTR(LOWER(location), '–ù–∏–∂–Ω–∏–π - –ù–æ–≤–≥–æ—Ä–æ–¥') > 0 THEN '–ù–∏–∂–Ω–∏–π –ù–æ–≤–≥–æ—Ä–æ–¥' \
            WHEN INSTR(LOWER(name), '–°–∞–º–∞—Ä–∞') > 0 OR INSTR(LOWER(location), '–°–∞–º–∞—Ä–∞') > 0 THEN '–°–∞–º–∞—Ä–∞' \
            WHEN INSTR(LOWER(name), '–ö–∞–∑–∞–Ω—å') > 0 OR INSTR(LOWER(location), '–ö–∞–∑–∞–Ω—å') > 0 THEN '–ö–∞–∑–∞–Ω—å' \
            WHEN INSTR(LOWER(name), '–û–º—Å–∫') > 0 OR INSTR(LOWER(location), '–û–º—Å–∫') > 0 THEN '–û–º—Å–∫' \
            WHEN INSTR(LOWER(name), '–ß–µ–ª—è–±–∏–Ω—Å–∫') > 0 OR INSTR(LOWER(location), '–ß–µ–ª—è–±–∏–Ω—Å–∫') > 0 THEN '–ß–µ–ª—è–±–∏–Ω—Å–∫' \
            WHEN INSTR(LOWER(name), '–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É') > 0 OR INSTR(LOWER(location), '–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É') > 0 THEN '–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É' \
            WHEN INSTR(LOWER(name), '–†–æ—Å—Ç–æ–≤ –Ω–∞ –î–æ–Ω—É') > 0 OR INSTR(LOWER(location), '–†–æ—Å—Ç–æ–≤ –Ω–∞ –î–æ–Ω—É') > 0 THEN '–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É' \
            WHEN INSTR(LOWER(name), '–†–æ—Å—Ç–æ–≤') > 0 OR INSTR(LOWER(location), '–†–æ—Å—Ç–æ–≤') > 0 THEN '–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É' \
            WHEN INSTR(LOWER(name), '–£—Ñ–∞') > 0 OR INSTR(LOWER(location), '–£—Ñ–∞') > 0 THEN '–£—Ñ–∞' \
            WHEN INSTR(LOWER(name), '–ü–µ—Ä–º—å') > 0 OR INSTR(LOWER(location), '–ü–µ—Ä–º—å') > 0 THEN '–ü–µ—Ä–º—å' \
            WHEN INSTR(LOWER(name), '–í–æ–ª–≥–æ–≥—Ä–∞–¥') > 0 OR INSTR(LOWER(location), '–í–æ–ª–≥–æ–≥—Ä–∞–¥') > 0 THEN '–í–æ–ª–≥–æ–≥—Ä–∞–¥' \
            WHEN INSTR(LOWER(name), '–ö—Ä–∞—Å–Ω–æ—è—Ä—Å–∫') > 0 OR INSTR(LOWER(location), '–ö—Ä–∞—Å–Ω–æ—è—Ä—Å–∫') > 0 THEN '–ö—Ä–∞—Å–Ω–æ—è—Ä—Å–∫' \
            WHEN INSTR(LOWER(name), '–í–æ—Ä–æ–Ω–µ–∂') > 0 OR INSTR(LOWER(location), '–í–æ—Ä–æ–Ω–µ–∂') > 0 THEN '–í–æ—Ä–æ–Ω–µ–∂' \
            WHEN INSTR(LOWER(name), '–°–∞—Ä–∞—Ç–æ–≤') > 0 OR INSTR(LOWER(location), '–°–∞—Ä–∞—Ç–æ–≤') > 0 THEN '–°–∞—Ä–∞—Ç–æ–≤' \
            WHEN INSTR(LOWER(name), '–¢–æ–ª—å—è—Ç—Ç–∏') > 0 OR INSTR(LOWER(location), '–¢–æ–ª—å—è—Ç—Ç–∏') > 0 THEN '–¢–æ–ª—å—è—Ç—Ç–∏' \
            WHEN INSTR(LOWER(name), '–¢–æ–ª—å—è—Ç–∏') > 0 OR INSTR(LOWER(location), '–¢–æ–ª—å—è—Ç–∏') > 0 THEN '–¢–æ–ª—å—è—Ç—Ç–∏' \
            WHEN INSTR(LOWER(name), '–ö—Ä–∞—Å–Ω–æ–¥–∞—Ä') > 0 OR INSTR(LOWER(location), '–ö—Ä–∞—Å–Ω–æ–¥–∞—Ä') > 0 THEN '–ö—Ä–∞—Å–Ω–æ–¥–∞—Ä' \
            WHEN INSTR(LOWER(name), '–ò–∂–µ–≤—Å–∫') > 0 OR INSTR(LOWER(location), '–ò–∂–µ–≤—Å–∫') > 0 THEN '–ò–∂–µ–≤—Å–∫' \
            WHEN INSTR(LOWER(name), '–Ø—Ä–æ—Å–ª–∞–≤–ª—å') > 0 OR INSTR(LOWER(location), '–Ø—Ä–æ—Å–ª–∞–≤–ª—å') > 0 THEN '–Ø—Ä–æ—Å–ª–∞–≤–ª—å' \
            WHEN INSTR(LOWER(name), '–£–ª—å—è–Ω–æ–≤—Å–∫') > 0 OR INSTR(LOWER(location), '–£–ª—å—è–Ω–æ–≤—Å–∫') > 0 THEN '–£–ª—å—è–Ω–æ–≤—Å–∫' \
            WHEN INSTR(LOWER(name), '–ë–∞—Ä–Ω–∞—É–ª') > 0 OR INSTR(LOWER(location), '–ë–∞—Ä–Ω–∞—É–ª') > 0 THEN '–ë–∞—Ä–Ω–∞—É–ª' \
            WHEN INSTR(LOWER(name), '–í–ª–∞–¥–∏–≤–æ—Å—Ç–æ–∫') > 0 OR INSTR(LOWER(location), '–í–ª–∞–¥–∏–≤–æ—Å—Ç–æ–∫') > 0 THEN '–í–ª–∞–¥–∏–≤–æ—Å—Ç–æ–∫' \
            WHEN INSTR(LOWER(name), '–•–∞–±–∞—Ä–æ–≤—Å–∫') > 0 OR INSTR(LOWER(location), '–•–∞–±–∞—Ä–æ–≤—Å–∫') > 0 THEN '–•–∞–±–∞—Ä–æ–≤—Å–∫' \
            WHEN INSTR(LOWER(name), '–ò—Ä–∫—É—Ç—Å–∫') > 0 OR INSTR(LOWER(location), '–ò—Ä–∫—É—Ç—Å–∫') > 0 THEN '–ò—Ä–∫—É—Ç—Å–∫' \
            WHEN INSTR(LOWER(name), '–¢—é–º–µ–Ω—å') > 0 OR INSTR(LOWER(location), '–¢—é–º–µ–Ω—å') > 0 THEN '–¢—é–º–µ–Ω—å' \
            WHEN INSTR(LOWER(name), '–ù–æ–≤–æ–∫—É–∑–Ω–µ—Ü–∫') > 0 OR INSTR(LOWER(location), '–ù–æ–≤–æ–∫—É–∑–Ω–µ—Ü–∫') > 0 THEN '–ù–æ–≤–æ–∫—É–∑–Ω–µ—Ü–∫' \
            WHEN INSTR(LOWER(name), '–û—Ä–µ–Ω–±—É—Ä–≥') > 0 OR INSTR(LOWER(location), '–û—Ä–µ–Ω–±—É—Ä–≥') > 0 THEN '–û—Ä–µ–Ω–±—É—Ä–≥' \
            WHEN INSTR(LOWER(name), '–ö–µ–º–µ—Ä–æ–≤–æ') > 0 OR INSTR(LOWER(location), '–ö–µ–º–µ—Ä–æ–≤–æ') > 0 THEN '–ö–µ–º–µ—Ä–æ–≤–æ' \
            WHEN INSTR(LOWER(name), '–†—è–∑–∞–Ω—å') > 0 OR INSTR(LOWER(location), '–†—è–∑–∞–Ω—å') > 0 THEN '–†—è–∑–∞–Ω—å' \
            WHEN INSTR(LOWER(name), '–ù–∞–±–µ—Ä–µ–∂–Ω—ã–µ –ß–µ–ª–Ω—ã') > 0 OR INSTR(LOWER(location), '–ù–∞–±–µ—Ä–µ–∂–Ω—ã–µ –ß–µ–ª–Ω—ã') > 0 THEN '–ù–∞–±–µ—Ä–µ–∂–Ω—ã–µ –ß–µ–ª–Ω—ã' \
            WHEN INSTR(LOWER(name), '–ù–∞–±–µ—Ä–µ–∂–Ω—ã–µ') > 0 OR INSTR(LOWER(location), '–ù–∞–±–µ—Ä–µ–∂–Ω—ã–µ') > 0 THEN '–ù–∞–±–µ—Ä–µ–∂–Ω—ã–µ –ß–µ–ª–Ω—ã' \
            WHEN INSTR(LOWER(name), '–ü–µ–Ω–∑–∞') > 0 OR INSTR(LOWER(location), '–ü–µ–Ω–∑–∞') > 0 THEN '–ü–µ–Ω–∑–∞' \
            WHEN INSTR(LOWER(name), '–ê—Å—Ç—Ä–∞—Ö–∞–Ω—å') > 0 OR INSTR(LOWER(location), '–ê—Å—Ç—Ä–∞—Ö–∞–Ω—å') > 0 THEN '–ê—Å—Ç—Ä–∞—Ö–∞–Ω—å' \
            WHEN INSTR(LOWER(name), '–õ–∏–ø–µ—Ü–∫') > 0 OR INSTR(LOWER(location), '–õ–∏–ø–µ—Ü–∫') > 0 THEN '–õ–∏–ø–µ—Ü–∫' \
           WHEN INSTR(LOWER(name), '–¢–æ–º—Å–∫') > 0 OR INSTR(LOWER(location), '–¢–æ–º—Å–∫') > 0 THEN '–¢–æ–º—Å–∫' \
            WHEN INSTR(LOWER(name), '–°–µ–≤–∞—Å—Ç–æ–ø–æ–ª—å') > 0 OR INSTR(LOWER(location), '–°–µ–≤–∞—Å—Ç–æ–ø–æ–ª—å') > 0 THEN '–°–µ–≤–∞—Å—Ç–æ–ø–æ–ª—å' \
            WHEN INSTR(LOWER(name), '–°—Ç–∞–≤—Ä–æ–ø–æ–ª—å') > 0 OR INSTR(LOWER(location), '–°—Ç–∞–≤—Ä–æ–ø–æ–ª—å') > 0 THEN '–°—Ç–∞–≤—Ä–æ–ø–æ–ª—å' \
            WHEN INSTR(LOWER(name), '–ë–∞–ª–∞—à–∏—Ö–∞') > 0 OR INSTR(LOWER(location), '–ë–∞–ª–∞—à–∏—Ö–∞') > 0 THEN '–ë–∞–ª–∞—à–∏—Ö–∞' \
            WHEN INSTR(LOWER(name), '–†—è–∑–∞–Ω—å') > 0 OR INSTR(LOWER(location), '–†—è–∑–∞–Ω—å') > 0 THEN '–†—è–∑–∞–Ω—å' \
            WHEN INSTR(LOWER(name), '–ú–∞—Ö–∞—á–∫–∞–ª–∞') > 0 OR INSTR(LOWER(location), '–ú–∞—Ö–∞—á–∫–∞–ª–∞') > 0 THEN '–ú–∞—Ö–∞—á–∫–∞–ª–∞' \
            WHEN INSTR(LOWER(name), '–®—ã–∫–º–µ–Ω—Ç') > 0 OR INSTR(LOWER(location), '–®—ã–∫–º–µ–Ω—Ç') > 0 THEN '–®—ã–º–∫–µ–Ω—Ç' \
            WHEN INSTR(LOWER(name), '–®—ã–º–∫–µ–Ω—Ç') > 0 OR INSTR(LOWER(location), '–®—ã–º–∫–µ–Ω—Ç') > 0 THEN '–®—ã–º–∫–µ–Ω—Ç' \
            WHEN INSTR(LOWER(name), '–ê–ª–º–∞-–ê—Ç–∞') > 0 OR INSTR(LOWER(location), '–ê–ª–º–∞-–ê—Ç–∞') > 0 THEN '–ê–ª–º–∞-–ê—Ç–∞' \
            WHEN INSTR(LOWER(name), '–ê–ª–º–∞ –ê—Ç–∞') > 0 OR INSTR(LOWER(location), '–ê–ª–º–∞ –ê—Ç–∞') > 0 THEN '–ê–ª–º–∞-–ê—Ç–∞' \
            WHEN INSTR(LOWER(name), '–ê–ª–º–∞ –ê—Ç—ã') > 0 OR INSTR(LOWER(location), '–ê–ª–º–∞ –ê—Ç—ã') > 0 THEN '–ê–ª–º–∞-–ê—Ç–∞' \
            WHEN INSTR(LOWER(name), '–ê–ª–º–∞-–ê—Ç—ã') > 0 OR INSTR(LOWER(location), '–ê–ª–º–∞-–ê—Ç—ã') > 0 THEN '–ê–ª–º–∞-–ê—Ç–∞' \
            WHEN INSTR(LOWER(name), '–ê—Å—Ç–∞–Ω–∞') > 0 OR INSTR(LOWER(location), '–ê—Å—Ç–∞–Ω–∞') > 0 THEN '–ê—Å—Ç–∞–Ω–∞' \
            WHEN INSTR(LOWER(name), '–ê–∫—Ç–æ–±–µ') > 0 OR INSTR(LOWER(location), '–ê–∫—Ç–æ–±–µ') > 0 THEN '–ê–∫—Ç–æ–±–µ' \
            WHEN INSTR(LOWER(name), '–ö–∞—Ä–∞–≥–∞–Ω–¥–∞') > 0 OR INSTR(LOWER(location), '–ö–∞—Ä–∞–≥–∞–Ω–¥–∞') > 0 THEN '–ö–∞—Ä–∞–≥–∞–Ω–¥–∞' \
           WHEN INSTR(LOWER(name), '–ú–∏–Ω—Å–∫') > 0 OR INSTR(LOWER(location), '–ú–∏–Ω—Å–∫') > 0 THEN '–ú–∏–Ω—Å–∫' \
            WHEN INSTR(LOWER(name), '–ì–æ–º–µ–ª—å') > 0 OR INSTR(LOWER(location), '–ì–æ–º–µ–ª—å') > 0 THEN '–ì–æ–º–µ–ª—å' \
            WHEN INSTR(LOWER(name), '–ì—Ä–æ–¥–Ω–æ') > 0 OR INSTR(LOWER(location), '–ì—Ä–æ–¥–Ω–æ') > 0 THEN '–ì—Ä–æ–¥–Ω–æ' \
            ELSE '–î—Ä—É–≥–æ–µ' \
       END AS city, \
       withSpacePass AS pass_status \
FROM `krugi-prod-social`.activity \
  -- –∏—Å–∫–ª—é—á–∏–º —Ç–µ—Å—Ç–æ–≤—ã–µ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è \
AND NOT (CHAR_LENGTH(TRIM(name)) BETWEEN 1 AND 4 \
                   OR LOWER(name) REGEXP '\\b—Ç–µ—Å—Ç[a-z–∞-—è—ë0-9 &@]*\\b|\\btest[a-z0-9 &@]*\\b') \
AND authorId NOT IN (SELECT id AS authorId \
                       FROM `krugi-prod-social`.users \
                      WHERE phone IN ('70333377777', \
                                      '12640826663333', \
                                      '72828203554', \
                                      '79160631144', \
                                      '78005553535', \
                                      '79619919840', \
                                      '79833856638', \
                                      '79831063666', \
                                      '79635088274', \
                                      '71111111111')) \
AND status = 'ACTIVE' \
AND deletedAt IS NULL" \

# –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –∑–∞–ø—Ä–æ—Å–∞ –∏ –∑–∞–≥—Ä—É–∑–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –≤ DataFrame
events = pd.read_sql(query, engine)
events['start_date'] = pd.to_datetime(events['start_date'], format='%d-%m-%Y')

# –í–ê–ñ–ù–û: –∏–∑ MySQL –ø—Ä–∏—Ö–æ–¥–∏—Ç —Ñ–æ—Ä–º–∞—Ç YYYY-MM-DD ‚Üí –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –º—è–≥–∫–æ
events['start_date'] = pd.to_datetime(events['start_date'], errors='coerce')
events['end_date']   = pd.to_datetime(events['end_date'], errors='coerce')

# ------------------------------------------------
# 1.2) –í—Å–µ–≥–æ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π (–∫–∞–∫ –∏ –±—ã–ª–æ —É —Ç–µ–±—è)
# ------------------------------------------------
engine_1 = create_engine('mysql+mysqlconnector://come_on_dude')
query_1 = """
SELECT COUNT(id)
FROM `krugi-prod-social`.activity
WHERE authorId NOT IN (SELECT id AS authorId
                         FROM `krugi-prod-social`.users
                        WHERE phone IN ('70333377777',
                                        '12640826663333',
                                        '72828203554',
                                        '79160631144',
                                        '78005553535',
                                        '79619919840',
                                        '79833856638',
                                        '79831063666',
                                        '79635088274',
                                        '71111111111'))
  AND deletedAt IS NULL
"""
events_all = pd.read_sql(query_1, engine_1)

# ------------------------------------------------
# 2) –§–∏–ª—å—Ç—Ä—ã –ø–æ –¥–∞—Ç–∞–º –∏ —Å—Ç–∞—Ç—É—Å–∞–º (–∫–∞–∫ —É —Ç–µ–±—è)
# ------------------------------------------------
# –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ first_date / second_date –∑–∞–¥–∞–Ω—ã –≤—ã—à–µ –∫–∞–∫ –¥–∞—Ç—ã/—Å—Ç—Ä–æ–∫–∏
events = events.query("@first_date <= start_date <= @second_date")
events = events.query("status == 'ACTIVE'")

# ------------------------------------------------
# 3) KPI-—Ç–∞–±–ª–∏—Ü–∞ (–∫–∞–∫ —É —Ç–µ–±—è)
# ------------------------------------------------
def kpis_events_week(events_week: pd.DataFrame) -> pd.DataFrame:
    df = events_week.copy()

    for col in ["status", "format"]:
        if col in df.columns and df[col].dtype == "O":
            df[col] = df[col].astype(str).str.upper().str.strip()

    df["pass_status"] = pd.to_numeric(df.get("pass_status", 0), errors="coerce").fillna(0).astype(int)

    dfa = df[df["status"] == "ACTIVE"].copy()

    is_online  = dfa["format"] == "ONLINE"
    is_offline = dfa["format"] == "OFFLINE"
    is_pass    = dfa["pass_status"] == 1

    total_events        = int(len(dfa))
    online_total        = int(is_online.sum())
    offline_total       = int(is_offline.sum())
    online_pass_total   = int((is_online & is_pass).sum())
    offline_pass_total  = int((is_offline & is_pass).sum())

    if not dfa.empty and "start_date" in dfa.columns:
        min_date = dfa["start_date"].min().strftime("%d.%m.%Y")
        max_date = dfa["start_date"].max().strftime("%d.%m.%Y")
        booking_period = f"{min_date} ‚Äî {max_date}"
    else:
        booking_period = "-"

    return pd.DataFrame({
        "–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å": [
            "–í—Å–µ–≥–æ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π",
            "–û–Ω–ª–∞–π–Ω –≤—Å–µ–≥–æ",
            "–û—Ñ—Ñ–ª–∞–π–Ω –≤—Å–µ–≥–æ",
            "–û–Ω–ª–∞–π–Ω –ø–æ SPACE Pass",
            "–û—Ñ—Ñ–ª–∞–π–Ω –ø–æ SPACE Pass",
            "–ü–µ—Ä–∏–æ–¥ –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è"
        ],
        "–ó–Ω–∞—á–µ–Ω–∏–µ": [
            total_events,
            online_total,
            offline_total,
            online_pass_total,
            offline_pass_total,
            booking_period
        ]
    })

events_week   = events.copy()
events_summary = kpis_events_week(events_week)

event_list = (
    events.loc[:, ["start_date", "name"]]
          .dropna(subset=["start_date", "name"])
          .sort_values("start_date")
          .reset_index(drop=True)
)

# ------------------------------------------------------------
# 4) –î–û–ë–ê–í–õ–Ø–ï–ú: result_df (–ø–æ 3 —Å–ª—É—á–∞–π–Ω—ã—Ö —Å–æ–±—ã—Ç–∏—è –Ω–∞ –∫–∞–∂–¥—ã–π –¥–µ–Ω—å)
# ------------------------------------------------------------

# 4.1. –°–æ–±–∏—Ä–∞–µ–º events_html –∏–∑ events
events_html = events[['event_id', 'start_date', 'end_date', 'name', 'pass_status']].copy()
events_html["link"] = "https://app.elsspace.ru/activity/" + events_html["event_id"].astype(str)

# 4.2. –î–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –ø–æ-—Ä—É—Å—Å–∫–∏
weekdays_ru = {
    0: "–ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫",
    1: "–í—Ç–æ—Ä–Ω–∏–∫",
    2: "–°—Ä–µ–¥–∞",
    3: "–ß–µ—Ç–≤–µ—Ä–≥",
    4: "–ü—è—Ç–Ω–∏—Ü–∞",
    5: "–°—É–±–±–æ—Ç–∞",
    6: "–í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ"
}
events_html["weekday"] = events_html["start_date"].dt.weekday.map(weekdays_ru)

# 4.3. –§–∏–ª—å—Ç—Ä—É–µ–º —è–≤–Ω—ã–µ "—Ç–µ—Å—Ç—ã" (–Ω–∞ —Å–ª—É—á–∞–π, –µ—Å–ª–∏ —á—Ç–æ-—Ç–æ –ø—Ä–æ—à–ª–æ –∏–∑ SQL)
df = events_html.copy()
df = df[~df["name"].str.contains("—Ç–µ—Å—Ç", case=False, na=False)]

# 4.4. –ö—Ä–∞—Å–∏–≤—ã–π –ø–æ—Ä—è–¥–æ–∫ –¥–Ω–µ–π
wd_order = ["–ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫","–í—Ç–æ—Ä–Ω–∏–∫","–°—Ä–µ–¥–∞","–ß–µ—Ç–≤–µ—Ä–≥","–ü—è—Ç–Ω–∏—Ü–∞","–°—É–±–±–æ—Ç–∞","–í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ"]
df["weekday"] = pd.Categorical(df["weekday"], categories=wd_order, ordered=True)

# 4.5. –°–ª—É—á–∞–π–Ω–∞—è –≤—ã–±–æ—Ä–∫–∞ –¥–æ 3 –Ω–∞ –¥–µ–Ω—å –±–µ–∑ .apply (–±–µ–∑ –≤–æ—Ä–Ω–∏–Ω–≥–æ–≤)
rng = np.random.RandomState(42)
df["_rand"] = rng.rand(len(df))
sampled = (
    df.sort_values(["weekday", "_rand"])
      .groupby("weekday", observed=False, group_keys=False)
      .head(3)
      .drop(columns="_rand")
)

# 4.6. –ì–æ—Ç–æ–≤–∏–º –æ—Ç—Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –ø–æ–ª–µ –∏ —Ñ–∏–Ω–∞–ª—å–Ω—ã–π DF
sampled["formatted"] = sampled.apply(
    lambda r: f"{r['weekday']}: ¬´{r['name']}¬ª, {r['link']}",
    axis=1
)
result_df = sampled[["weekday", "name", "link", "formatted"]].reset_index(drop=True)

# ------------------------------------------------------------
# 5) –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã–µ —Ñ–æ—Ä–º–∞—Ç—Ç–µ—Ä—ã (–¥–æ–ø–æ–ª–Ω–∏–º –¥–ª—è result_df)
# ------------------------------------------------------------
def _format_events_summary(df: pd.DataFrame) -> str:
    lines = []
    for _, row in df.iterrows():
        metric = str(row.get("–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å", "")).strip()
        value  = str(row.get("–ó–Ω–∞—á–µ–Ω–∏–µ", "")).strip()
        line = f"‚Äî {metric} {value}" if metric.endswith(":") else f"‚Äî {metric}: {value}"
        lines.append(line)
    return "\n".join(lines) if lines else "‚Äî (–Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö)"

def _format_event_list(df: pd.DataFrame, date_fmt: str = "%d.%m") -> str:
    out = []
    for _, row in df.iterrows():
        dt = pd.to_datetime(row["start_date"], errors="coerce")
        ds = "" if pd.isna(dt) else dt.strftime(date_fmt)
        name = str(row["name"]).strip()
        out.append(f"{ds} ‚Äî {name}" if ds else name)
    return "\n".join(out) if out else "–°–ø–∏—Å–æ–∫ –ø–æ–∫–∞ –ø—É—Å—Ç."

def _format_events_all(events_all) -> str:
    if isinstance(events_all, (int, float)):
        total = int(events_all)
    elif isinstance(events_all, pd.DataFrame):
        if events_all.size == 1:
            total = int(pd.to_numeric(events_all.stack().iloc[0], errors="coerce"))
        else:
            for col in ["–ó–Ω–∞—á–µ–Ω–∏–µ", "value", "total", "count"]:
                if col in events_all.columns:
                    total = int(pd.to_numeric(events_all[col].iloc[0], errors="coerce"))
                    break
            else:
                total = int(pd.to_numeric(events_all.select_dtypes(include="number").stack().iloc[0], errors="coerce"))
    else:
        raise ValueError("events_all –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —á–∏—Å–ª–æ–º –∏–ª–∏ DataFrame —Å –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–º —á–∏—Å–ª–æ–º.")
    return f"–ú–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π –∑–∞ –≤—Å–µ –≤—Ä–µ–º—è: {total}"

# –ù–û–í–û–ï: —Ñ–æ—Ä–º–∞—Ç—Ç–µ—Ä –¥–ª—è –±–ª–æ–∫–∞ "–ú–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è –Ω–∞ –∫–∞–∂–¥—ã–π –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏" –∏–∑ result_df
def _format_events_by_weekday_block(result_df: pd.DataFrame) -> str:
    if result_df.empty:
        return "‚Äî (–Ω–∞ —ç—Ç–æ–π –Ω–µ–¥–µ–ª–µ –Ω–µ—Ç –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö —Å–æ–±—ã—Ç–∏–π)"
    # –ø—Ä–æ—Å—Ç–æ –≤—ã–≤–æ–¥–∏–º –∫–∞–∂–¥—É—é —Å—Ç—Ä–æ–∫—É –∏–∑ 'formatted'
    return "\n".join(result_df["formatted"].tolist())

# ------------------------------------------------------------
# 6) –†–µ–Ω–¥–µ—Ä —Ç–µ–∫—Å—Ç–∞: –¥–æ–±–∞–≤–∏–º –≤—Å—Ç–∞–≤–∫—É result_df –ø–æ –º–∞—Ä–∫–µ—Ä—É
# ------------------------------------------------------------
def render_post_with_data(
    template_text: str,
    events_summary: pd.DataFrame,
    event_list: pd.DataFrame,
    events_all,               # —á–∏—Å–ª–æ –∏–ª–∏ DataFrame
    result_df: pd.DataFrame,  # –ù–û–í–´–ô –∞—Ä–≥—É–º–µ–Ω—Ç
    summary_marker_line: str = "–í—Å—Ç–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ events_summary",
    list_marker_line:    str = "–í—Å—Ç–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ event_list",
    all_marker_inline:   str = "–í—Å—Ç–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ events_all",
    result_marker_line:  str = "–í—Å—Ç–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ result_df",  # –ù–û–í–´–ô –º–∞—Ä–∫–µ—Ä
    date_fmt: str = "%d.%m"
) -> str:
    summary_block = _format_events_summary(events_summary)
    list_block    = _format_event_list(event_list, date_fmt=date_fmt)
    all_block     = _format_events_all(events_all)
    result_block  = _format_events_by_weekday_block(result_df)  # –≥–æ—Ç–æ–≤–∏–º –±–ª–æ–∫ –¥–ª—è result_df

    rendered_lines = []
    for line in template_text.splitlines():
        s = line.strip()
        if s == summary_marker_line:
            rendered_lines.append(summary_block)
        elif s == list_marker_line:
            rendered_lines.append(list_block)
        elif s == result_marker_line:                     # –≤—Å—Ç–∞–≤–∫–∞ –ø–æ –Ω–æ–≤–æ–º—É –º–∞—Ä–∫–µ—Ä—É
            rendered_lines.append(result_block)
        else:
            if all_marker_inline in line:
                line = line.replace(all_marker_inline, all_block)
            rendered_lines.append(line)
    return "\n".join(rendered_lines).strip()

# ------------------------------------------------------------
# 7) –ò—Å–ø–æ–ª—å–∑—É–µ–º
# ------------------------------------------------------------
template_text = """–ó–∞–¥–∞—á–∞:
–ù–∞–ø–∏—à–∏ –ø–æ—Å—Ç –¥–ª—è —Å–æ—Ü–∏–∞–ª—å–Ω–æ–π —Å–µ—Ç–∏ SPACE, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Ä–∞–∑–º–µ—â–∞—Ç—å—Å—è –≤ –Ω–∞—á–∞–ª–µ –Ω–µ–¥–µ–ª–∏ –∏ –º–æ—Ç–∏–≤–∏—Ä–æ–≤–∞—Ç—å –ª—é–¥–µ–π –∑–∞–ø–∏—Å–∞—Ç—å—Å—è –Ω–∞ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è.

–°—Ç—Ä—É–∫—Ç—É—Ä–∞ –ø–æ—Å—Ç–∞:
–≠–º–æ—Ü–∏–æ–Ω–∞–ª—å–Ω—ã–π –∑–∞—Ö–æ–¥ ‚Äî –≤–¥–æ—Ö–Ω–æ–≤–ª—è—é—â–∏–π –∏ –º–æ—Ç–∏–≤–∏—Ä—É—é—â–∏–π, –æ—Ç –ª–∏—Ü–∞ —Å–æ—á—É–≤—Å—Ç–≤—É—é—â–µ–≥–æ –¥—Ä—É–≥–∞. –≠—Ç–æ –ø–æ—Å—Ç –æ—Ç–∫—Ä—ã–≤–∞—é—â–∏–π –Ω–µ–¥–µ–ª—é. 
–í—Å—Ç—É–ø–ª–µ–Ω–∏–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —ç–Ω–µ—Ä–≥–∏—á–Ω—ã–º, –º–æ—Ç–∏–≤–∏—Ä—É—é—â–∏–º, –∑–∞—Ä—è–∂–∞—é—â–∏–º —ç–Ω–µ—Ä–≥–∏–µ–π –∏ –±—ã—Ç—å –æ—Å–Ω–æ–≤–∞–Ω–æ –ø–æ —Å–º—ã—Å–ª—É –Ω–∞ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è—Ö —ç—Ç–æ–π –Ω–µ–¥–µ–ª–∏. 
–ò –¥–æ–ª–∂–Ω–æ —Å–æ–¥–µ—Ä–∂–∞—Ç—å –ø–µ—Ä–µ—Ö–æ–¥ –Ω–∞ —Ç–æ, —á—Ç–æ –∫–∞–∫–∏–µ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è —Ç–µ–±–µ –¥–æ—Å—Ç—É–ø–Ω—ã –Ω–∞ —ç—Ç–æ–π –Ω–µ–¥–µ–ª–µ –≤ SPACE –∏ –ø–µ—Ä–µ–π—Ç–∏ –∫ –∞–≥—Ä–µ–≥–∏—Ä—É—é—â–∏–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è–º –∏ –¥–∞–ª–µ–µ

–ê–≥—Ä–µ–≥–∏—Ä—É—é—â–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –Ω–µ–¥–µ–ª–∏ ‚Äî –¢—É—Ç —Ä–∞—Å—Å–∫–∞–∂–∏, —á—Ç–æ –Ω–∞ —ç—Ç–æ–π –Ω–µ–¥–µ–ª–µ —Ç—ã –º–æ–∂–µ—à—å –ø–æ—Å–µ—Ç–∏—Ç—å —Å—Ç–æ–ª—å–∫–æ-—Ç–æ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π, —Ç.–µ —Ä–∞—Å—Å–∫–∞–∂–∏ –≤ —Ç–µ–∫—Å—Ç–µ.
–ò—Å–ø–æ–ª—å–∑—É–π –¥–ª—è —ç—Ç–æ–≥–æ –∞–≥—Ä–µ–≥–∏—Ä—É—é—â–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏, –∫–æ—Ç–æ—Ä—ã–µ —É–∫–∞–∑–∞–Ω—ã –≤ ‚Äú–ê–≥—Ä–µ–≥–∏—Ä—É—é—â–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏‚Äù. 

–í —ç—Ç–æ–º –±–ª–æ–∫–µ –æ—Ç–¥–µ–ª—å–Ω–æ —Ä–∞—Å—Å–º–æ—Ç—Ä–∏ –≤—Å–µ, —á—Ç–æ –∫–∞—Å–∞–µ—Ç—Å—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π SPACE Pass, —Å–∫–∞–∑–∞–≤, —á—Ç–æ –µ—Å–ª–∏ —É —Ç–µ–±—è –µ—Å—Ç—å –∫–∞—Ä—Ç–∞, —Ç–æ –≤–æ–æ–±—â–µ –∫—Ä—É—Ç–æ. 

–î–∞–ª–µ–µ —Ä–∞—Å—Å–∫–∞–∂–∏ –æ —Ç–æ–º, —Å–∫–æ–ª—å–∫–æ –≤—Å–µ–≥–æ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π –∑–∞ –≤—Å–µ –≤—Ä–µ–º—è –±—ã–ª–æ ‚Äî –í—Å—Ç–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ events_all

–ê–Ω–æ–Ω—Å—ã –ø–æ –¥–Ω—è–º ‚Äî –∫–∞–∂–¥—ã–π –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –æ–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –≤ 1‚Äì2 –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è—Ö.
- –ù–µ –¥–µ–ª–∞–π —Å–ø–∏—Å–∫–∏ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π.
- –°—Ç–∏–ª—å: ¬´–í —ç—Ç–æ—Ç –¥–µ–Ω—å —Ç—ã –º–æ–∂–µ—à—å‚Ä¶¬ª, ¬´–¢—ã —É–∑–Ω–∞–µ—à—å‚Ä¶¬ª, ¬´–≠—Ç–æ –ø–æ–º–æ–∂–µ—Ç‚Ä¶¬ª
- –û–±—â–∏–π –æ–±–∑–æ—Ä, –∞ –Ω–µ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–∏–µ.
- –û–±—ä—ë–º –∫–∞–∂–¥–æ–≥–æ –∞–Ω–æ–Ω—Å–∞ ‚Äî –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –∫—Ä–∞—Ç–∫–∏–π –∏ —ë–º–∫–∏–π (2 —Ä–∞–∑–∞ –∫–æ—Ä–æ—á–µ –æ–±—ã—á–Ω–æ–≥–æ —Ç–µ–∫—Å—Ç–∞).
- –û–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –ø–æ–∫–∞–∑–∞—Ç—å, —á–µ–º –ø–æ–ª–µ–∑–Ω—ã —Å–æ–±—ã—Ç–∏—è –¥–Ω—è (–∑–¥–æ—Ä–æ–≤—å–µ, –≤–¥–æ—Ö–Ω–æ–≤–µ–Ω–∏–µ, –±–∏–∑–Ω–µ—Å, –æ–±—â–µ–Ω–∏–µ). 
–ò –¥–æ–±–∞–≤—å —Ç—Ä–∏ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è –≤ –∫–∞–∂–¥—ã–π –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏.–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –¥–ª—è —ç—Ç–æ–≥–æ –±–ª–æ–∫–∞ –≤–æ–∑—å–º–∏ –≤ —Ä–∞–∑–¥–µ–ª–µ "–ú–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è –Ω–∞ –∫–∞–∂–¥—ã–π –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏:" –ù–∞–∑–≤–∞–Ω–∏–µ –∏ —Å—Å—ã–ª–∫–∞. –ù–∞–∑–≤–∞–Ω–∏–µ –≤ –∫–∞–≤—ã—á–∫–∞—Ö ¬´¬ª
–°–∫–∞–∑–∞–≤, —á—Ç–æ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –¥–Ω—è –≤—ã–±—Ä–∞–ª–∏ –ò–ò SPACE –≤—ã–±—Ä–∞–ª —Ç—Ä–∏ —Å–ª—É—á–∞–π–Ω—ã—Ö –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è. –ö–∞–∂–¥–æ–µ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–¢.–µ –∫–∞–∫ –ø—Ä–∏–º–µ—Ä, –≤–æ—Ç –≤ —Ç–∞–∫–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä–µ:

–ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫ ‚Äî –º—è–≥–∫–∞—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –∏ –≤–Ω—É—Ç—Ä–µ–Ω–Ω—è—è –æ–ø–æ—Ä–∞

–í —ç—Ç–æ—Ç –¥–µ–Ω—å —Ç—ã –º–æ–∂–µ—à—å –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —ç–Ω–µ—Ä–≥–∏—é, —É—Å–ª—ã—à–∞—Ç—å —Ç–µ–ª–æ –∏ –≤–µ—Ä–Ω—É—Ç—å —Ñ–æ–∫—É—Å. –≠—Ç–æ –ø–æ–º–æ–∂–µ—Ç –≤–æ–π—Ç–∏ –≤ –Ω–µ–¥–µ–ª—é —Å–ø–æ–∫–æ–π–Ω–æ –∏ –æ—Å–æ–∑–Ω–∞–Ω–Ω–æ.

‚Ä¢ –î–ï–¢–û–ö–° —Å–æ –°–º—ã—Å–ª–æ–º ‚Äî https://app.elsspace.ru/activity/34528
‚Ä¢ –ö–æ–Ω—Å—É–ª—å—Ç–∞—Ü–∏—è ‚Äî https://app.elsspace.ru/activity/38213
‚Ä¢ –î–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∞ —Å–æ–º–∞—Ç–∏–ø–æ–ª–æ–≥–∞ ‚Äî https://app.elsspace.ru/activity/27813

–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ ‚Äî –º–æ—Ç–∏–≤–∞—Ü–∏—è –ø–µ—Ä–µ–π—Ç–∏ –≤ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ SPACE, –Ω–∞–π—Ç–∏ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è –∏ —Å–æ–±—ã—Ç–∏—è –∏ –≤—Å—Ç–∞–≤–∏—Ç—å —Å—Å—ã–ª–∫—É https://app.elsspace.ru/activities 
–í–∞–∂–Ω–æ ‚Äî –ø–µ—Ä–µ—Ö–æ–¥ –º–µ–∂–¥—É –±–ª–æ–∫–∞–º–∏ –≤ —Å—Ç—Ä—É–∫—Ç—É—Ä–µ –ø–æ—Å—Ç–∞ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –º—è–≥–∫–∏–π, –Ω–∞–ø—Ä–∏–º–µ—Ä, –ù–∞ —ç—Ç–æ–π –Ω–µ–¥–µ–ª–µ —É –Ω–∞—Å, –¥–∞–≤–∞–π –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ —É –Ω–∞—Å –±—É–¥–µ—Ç –≤ –∫–∞–∂–¥—ã–π –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏. –ú–µ–∂–¥—É –±–ª–æ–∫–∞–º–∏ –¥–µ–ª–∞–π –æ—Ç—Å—Ç—É–ø.


–ê–≥—Ä–µ–≥–∏—Ä—É—é—â–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏:
–í—Å—Ç–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ events_summary

–°–ø–∏—Å–æ–∫ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π:
–í—Å—Ç–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ event_list

–ú–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è –Ω–∞ –∫–∞–∂–¥—ã–π –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏:
–í—Å—Ç–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ result_df

–°—Ç–∏–ª—å –ø–æ—Å—Ç–∞:
 ‚Äì –¢—ë–ø–ª—ã–π, –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞—é—â–∏–π —Ç–æ–Ω, –æ—Ç –ª–∏—Ü–∞ —Å–æ—á—É–≤—Å—Ç–≤—É—é—â–µ–≥–æ –¥—Ä—É–≥–∞.
 ‚Äì –û–±—Ä–∞—â–µ–Ω–∏–µ –Ω–∞ ¬´—Ç—ã¬ª (–≥–µ–Ω–¥–µ—Ä–Ω–æ-–Ω–µ–π—Ç—Ä–∞–ª—å–Ω–æ–µ).
 ‚Äì –°—Ç—Ä—É–∫—Ç—É—Ä–∞: —ç–º–æ—Ü–∏–æ–Ω–∞–ª—å–Ω—ã–π –∑–∞—Ö–æ–¥ ‚Üí —Å–∏—Ç—É–∞—Ü–∏—è/–Ω–∞–±–ª—é–¥–µ–Ω–∏–µ/–≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π –¥–∏–∞–ª–æ–≥ ‚Üí —ç–º–ø–∞—Ç–∏—è ‚Üí —Ä–µ—à–µ–Ω–∏–µ ‚Üí –∑–∞—á–µ–º —ç—Ç–æ –≤ –∂–∏–∑–Ω–∏.
 ‚Äì –í –∫–æ–Ω—Ü–µ: –º–æ—Ç–∏–≤–∏—Ä—É–π –ø–µ—Ä–µ–π—Ç–∏ –≤ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ SPACE 

–≠–º–æ–¥–∂–∏:
 –ò—Å–ø–æ–ª—å–∑—É–π 2‚Äì3 –∏–∑ —Å–ø–∏—Å–∫–∞ ‚ò∫Ô∏è üôÉ üôè üëâ üéØ üî• ‚ù§Ô∏è ‚úÖ. –í—Å—Ç–∞–≤–ª—è–π –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ, –∞ –Ω–µ –ø–æ–¥—Ä—è–¥ –∏ –Ω–µ –∫–∞–∫ —É–∫—Ä–∞—à–µ–Ω–∏–µ.
–ì–æ–≤–æ—Ä–∏ –∫–∞–∫ —Å–æ—á—É–≤—Å—Ç–≤—É—é—â–∏–π –¥—Ä—É–≥. –í –∫–æ–Ω—Ü–µ –ø–æ—Å—Ç–∞ –º–æ—Ç–∏–≤–∏—Ä—É–π –ø–µ—Ä–µ–π—Ç–∏ –≤ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –µ–≥–æ, –Ω–∞–π—Ç–∏ –∫—Ä—É–≥–∏ –∏–ª–∏ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è. –ê —Ç–∞–∫–∂–µ –º–æ—Ç–∏–≤–∏—Ä—É–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –æ—Å—Ç–∞–≤—å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ –ø–æ—Å—Ç—É, –ø–æ–¥–µ–ª–∏—Ç—å—Å—è –æ–ø—ã—Ç–æ–º, —Å–æ–≤–µ—Ç–∞–º–∏ —Å –¥—Ä—É–≥–∏–º–∏

–ü—Ä–∏–º–µ—Ä–Ω–∞—è —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –ø–æ—Å—Ç–æ–≤, –Ω–∞ —á—Ç–æ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å—Å—è, –Ω–∏–∂–µ, —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ –ø—Ä–∏–º–µ—Ä—ã —Å—Ç—Ä—É–∫—Ç—É—Ä—ã, —Ç–µ–º–ø–∞ –ø–æ—Å—Ç–æ–≤, –∞ –Ω–µ —Ç–æ –æ —á–µ–º –ø–∏—Å–∞—Ç—å.: 
(–¥–∞–ª–µ–µ —Ç–≤–æ–π –¥–ª–∏–Ω–Ω—ã–π –ø—Ä–∏–º–µ—Ä –æ—Å—Ç–∞—ë—Ç—Å—è –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏–π)

–ü–æ—Å—Ç 1 
–ê –≤—ã —Å–ª—ã—à–∞–ª–∏ —Ç–∞–∫–æ–µ: ¬´–í–æ—Ç –µ—Å–ª–∏ –±—ã –º–µ–Ω—è –∑–∞—Å—Ç–∞–≤–∏–ª–∏ —É—á–∏—Ç—å –∞–Ω–≥–ª–∏–π—Å–∫–∏–π –≤ –¥–µ—Ç—Å—Ç–≤–µ, —è –±—ã —É–∂–µ –¥–∞–≤–Ω–æ —Ö–æ—Ä–æ—à–æ –≥–æ–≤–æ—Ä–∏–ª!¬ª ? –ü—Ä–µ–¥–ª–∞–≥–∞–µ–º –ø–æ—á–∏—Ç–∞—Ç—å –æ —Ç–æ–º, –∫–∞–∫ –ø–æ–º–æ—á—å —Ä–µ–±–µ–Ω–∫—É –≤—ã—É—á–∏—Ç—å —è–∑—ã–∫, –Ω–µ –∑–∞—Å—Ç–∞–≤–ª—è—è –µ–≥–æ. - –ò—â–∏—Ç–µ —Å–æ–≤—Ä–µ–º–µ–Ω–Ω—É—é —à–∫–æ–ª—É –∞–Ω–≥–ª–∏–π—Å–∫–æ–≥–æ —è–∑—ã–∫–∞. –ì–∞–¥–∂–µ—Ç—ã, —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–∏, –º–æ–±–∏–ª—å–Ω—ã–µ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è ‚Äì –ø—Ä–æ—á–Ω–æ –≤–æ—à–ª–∏ –≤ –∂–∏–∑–Ω—å —Å–æ–≤—Ä–µ–º–µ–Ω–Ω—ã—Ö –¥–µ—Ç–µ–π. ¬´–°–∫—É–∫–∞¬ª ‚Äî –æ–¥–Ω–æ –∏–∑ —á–∞—Å—Ç—ã—Ö —Å–ª–æ–≤, –∫–æ–≥–¥–∞ —Ä–µ–±–µ–Ω–æ–∫ –≥–æ–≤–æ—Ä–∏—Ç –æ–± –æ–±—É—á–µ–Ω–∏–∏. –ü–∞—Ä—Ç—ã, –º–µ–ª, –¥–æ—Å–∫–∞, —É–∫–∞–∑–∫–∞ ‚Äì –≤–∞–º —É–∂–µ —Å—Ç–∞–ª–æ —Å–∫—É—á–Ω–æ? - –ö—Ä—É–≥–æ–∑–æ—Ä –ø—Ä–µ–ø–æ–¥–∞–≤–∞—Ç–µ–ª—è. –í–∞–∂–Ω–æ, –∫–æ–≥–¥–∞ –ø–µ–¥–∞–≥–æ–≥ –≤ –∫—É—Ä—Å–µ —Ç–æ–≥–æ, —á–µ–º –∏–Ω—Ç–µ—Ä–µ—Å—É—é—Ç—Å—è –¥–µ—Ç–∏. –° –Ω–∏–º –º–æ–∂–Ω–æ –æ–±—Å—É–¥–∏—Ç—å ¬´–º–∞–π–Ω–∫—Ä–∞—Ñ—Ç¬ª –∏–ª–∏ –ø–æ—à—É—Ç–∏—Ç—å —Ü–∏—Ç–∞—Ç–æ–π –∏–∑ –∫–æ–º–∏–∫—Å–æ–≤. –®–∫–æ–ª—å–Ω–∏–∫—É –≤–∞–∂–Ω–æ –∑–Ω–∞—Ç—å, —á—Ç–æ —Å –Ω–∏–º —Ä—è–¥–æ–º –≤–∑—Ä–æ—Å–ª—ã–π –¥—Ä—É–≥. - –ö–æ–º—å—é–Ω–∏—Ç–∏. –ú–∞–ª–µ–Ω—å–∫–∞—è –∫–æ–º–ø–∞–Ω–∏—è –µ–¥–∏–Ω–æ–º—ã—à–ª–µ–Ω–Ω–∏–∫–æ–≤, —Ç–∞–∫–∏—Ö –∂–µ –º–∞–ª—å—á–∏—à–µ–∫ –∏ –¥–µ–≤—á–æ–Ω–æ–∫, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ—Å–ª–µ —É—Ä–æ–∫–∞ –º–æ–≥—É—Ç –ø–æ–±–æ–ª—Ç–∞—Ç—å –∏ –ø–æ–ø–∏—Ç—å —á–∞–π –≤–º–µ—Å—Ç–µ, –Ω–∏–∫—Ç–æ –∏—Ö –Ω–µ –ø–æ—Ç–æ—Ä–æ–ø–∏—Ç. –û–±—â–µ–Ω–∏–µ ‚Äì –≤–∞–∂–Ω–∞—è —á–∞—Å—Ç—å –æ–±—Ä–∞–∑–æ–≤–∞—Ç–µ–ª—å–Ω–æ–≥–æ –ø—Ä–æ—Ü–µ—Å—Å–∞. - –ú–æ—Ç–∏–≤–∞—Ü–∏—è. –ú–Ω–æ–≥–∏–µ —É—á—Ä–µ–∂–¥–µ–Ω–∏—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–≥–æ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –¥–∞—é—Ç –¥–µ—Ç—è–º –Ω–∞–∫–ª–µ–π–∫–∏/—Å—Ç–∏–∫–µ—Ä—ã/–∫–∞—Ä—Ç–æ—á–∫–∏ –∑–∞ —Ö–æ—Ä–æ—à—É—é —Ä–∞–±–æ—Ç—É –≤ –∫–ª–∞—Å—Å–µ –∏–ª–∏ –¥–æ–º–∞. –ò–Ω—Ç–µ—Ä–µ—Å—É–π—Ç–µ—Å—å –ø–æ–¥–∞—Ä–∫–∞–º–∏, —Å–ø—Ä–∞—à–∏–≤–∞–π—Ç–µ —É—Å–ª–æ–≤–∏—è –º–æ—Ç–∏–≤–∞—Ü–∏–∏. –ú–ª–∞–¥—à–∏—Ö —à–∫–æ–ª—å–Ω–∏–∫–æ–≤ –≥–µ–π–º–∏—Ñ–∏–∫–∞—Ü–∏—è —Å—Ç–∏–º—É–ª–∏—Ä—É–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å –¥–æ–º–∞—à–Ω–∏–µ –∑–∞–¥–∞–Ω–∏—è –∏ –∞–∫—Ç–∏–≤–Ω–æ —Ä–∞–±–æ—Ç–∞—Ç—å –Ω–∞ —É—Ä–æ–∫–µ. –ü–æ–º–Ω–∏—Ç–µ, —á—Ç–æ –æ—Ç –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–µ—Å—Ç–∞ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫ –æ–±—É—á–µ–Ω–∏—é —Ä–µ–±–µ–Ω–∫–∞ –≤ —Ü–µ–ª–æ–º ‚Äì —Å–∫—É—á–Ω–æ —É—á–∏—Ç—å—Å—è –∏–ª–∏ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ. –ü–µ—Ä–µ—É–±–µ–∂–¥–∞—Ç—å –µ–≥–æ —Å –∫–∞–∂–¥—ã–º —Ä–∞–∑–æ–º –±—É–¥–µ—Ç —Å–ª–æ–∂–Ω–µ–µ. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ, —á—Ç–æ –µ—Å—Ç—å –º–µ—Å—Ç–æ, –≥–¥–µ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ —É—á–∏—Ç—å –∞–Ω–≥–ª–∏–π—Å–∫–∏–π —è–∑—ã–∫, –º–æ–∂–Ω–æ –ø–æ—Å–µ—Ç–∏–≤ –ø—Ä–æ–±–Ω—ã–π —É—Ä–æ–∫. –ú—ã —Ç–∞–∫–∏–µ –ø—Ä–æ–≤–æ–¥–∏–º. 

–ü–æ—Å—Ç 2 
–û–∫, –≤–æ—Ç —Ä–µ–±–µ–Ω–æ–∫ —É—á–∏—Ç—Å—è. –ê –∫–∞–∫ –ø–æ–Ω—è—Ç—å-—Ç–æ —É—Å–ø–µ—à–Ω–æ –æ–Ω –ø—Ä–æ—Ö–æ–¥–∏—Ç –æ–±—É—á–µ–Ω–∏–µ –∏ –µ—Å—Ç—å –ª–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç? –ü–µ—Ä–≤–æ–µ, —á—Ç–æ –ø—Ä–∏—Ö–æ–¥–∏—Ç –Ω–∞ —É–º, —ç—Ç–æ –∫–æ–Ω–µ—á–Ω–æ –∂–µ –æ—Ü–µ–Ω–∫–∏ –≤ —à–∫–æ–ª–µ. –ù–û! –ó–¥–µ—Å—å –µ—Å—Ç—å –æ—á–µ–Ω—å –±–æ–ª—å—à–æ–π –ø–æ–¥–≤–æ—Ö!!! –ö–æ–Ω–µ—á–Ω–æ, –µ—Å–ª–∏ —Ä–µ–±—ë–Ω–æ–∫ –Ω–∞—á–∞–ª –∑–∞–Ω–∏–º–∞—Ç—å—Å—è —É –Ω–∞—Å –≤ –ø–µ—Ä–≤–æ–º, –∏–ª–∏ –≤–æ –≤—Ç–æ—Ä–æ–º –∫–ª–∞—Å—Å–µ, —Ç–æ –µ—Å—Ç—å –¥–æ —Ç–æ–≥–æ (–∏–ª–∏ –≤ –æ–¥–∏–Ω –≥–æ–¥) –∫–∞–∫ –Ω–∞—á–∞–ª–∏ —É—á–∏—Ç—å –∞–Ω–≥–ª–∏–π—Å–∫–∏–π –≤ —à–∫–æ–ª–µ, —Ç–æ –∑–¥–µ—Å—å –≤—Å–µ –æ—á–µ–≤–∏–¥–Ω–æ - —Ä–µ–±—ë–Ω–∫—É –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ —à–∫–æ–ª–µ –ª–µ–≥–∫–æ. –ù–æ –≤–æ—Ç –µ—Å–ª–∏ –∫ –Ω–∞–º –ø—Ä–∏—Ö–æ–¥–∏—Ç —É—á–µ–Ω–∏–∫, –∫–æ—Ç–æ—Ä—ã–π —É–∂–µ –¥–≤–∞, —Ç—Ä–∏, —á–µ—Ç—ã—Ä–µ –≥–æ–¥–∞ –∏–∑—É—á–∞–µ—Ç –∞–Ω–≥–ª–∏–π—Å–∫–∏–π –≤ —à–∫–æ–ª–µ, –∞ –µ–≥–æ —É—Ä–æ–≤–µ–Ω—å –ø—Ä–∏–º–µ—Ä–Ω–æ –Ω—É–ª–µ–≤–æ–π –†–µ–∞–ª—å–Ω—ã–π —Å–ª—É—á–∞–π –∏–∑ –ø—Ä–∞–∫—Ç–∏–∫–∏ –í —à–∫–æ–ª–µ –Ω–∏–∫—Ç–æ –Ω–∏–∫–æ–≥–æ –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ –Ω–µ –∂–¥—ë—Ç, –ø—Ä–æ–≥—Ä–∞–º–º–∞ —É—Å–ª–æ–∂–Ω—è–µ—Ç—Å—è, —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è –≤—ã—Å–æ–∫–∏–µ, –∞ —É—á–µ–Ω–∏–∫ –Ω–µ —Ç—è–Ω–µ—Ç... –û–Ω –ø—Ä–∏—Ö–æ–¥–∏—Ç –∫ –Ω–∞–º, –ø—Ä–æ—Ö–æ–¥–∏—Ç —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –≥—Ä—É–ø–ø—É —Å –Ω–∞—á–∞–ª—å–Ω—ã–º (–µ–≥–æ) —É—Ä–æ–≤–Ω–µ–º. –ù–æ –∑–∞ —Ç—Ä–∏ –º–µ—Å—è—Ü–∞ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ —É—Å–≤–æ–∏—Ç—å –º–∞—Ç–µ—Ä–∏–∞–ª —Ç—Ä—ë—Ö –ª–µ—Ç –∏ –æ—Ü–µ–Ω–∫–∏ –≤ —à–∫–æ–ª–µ –æ—Å—Ç–∞—é—Ç—Å—è –ø—Ä–µ–∂–Ω–∏–º–∏ (–ø–æ–∫–∞!!!), –∞ –ø—Ä–æ–≥—Ä–µ—Å—Å —É—á–µ–Ω–∏–∫–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–º –Ω–µ —Å—Ç–æ–∏—Ç —Ç–∞–∫–∂–µ –∑–∞–±—ã–≤–∞—Ç—å, —á—Ç–æ —à–∫–æ–ª—å–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ - —ç—Ç–æ –≤—Å–µ–≥–¥–∞ —Å—É–±—ä–µ–∫—Ç–∏–≤–Ω–æ –ò –≤–æ—Ç –∑–¥–µ—Å—å –ø—Ä–∏—Ö–æ–¥–∏—Ç –Ω–∞ –ø–æ–º–æ—â—å –∞–±—Å–æ–ª—é—Ç–Ω–æ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –±—Ä–∏—Ç–∞–Ω—Å–∫–∏–µ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –∫–æ—Ç–∏—Ä—É—é—Ç—Å—è –≤–æ –≤—Å–µ–º –º–∏—Ä–µ –ê —à–∫–æ–ª—É –æ–Ω –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –Ω–∞–≥–æ–Ω–∏—Ç –∏ –ø–µ—Ä–µ–≥–æ–Ω–∏—Ç, –∏ —Ç–æ—á–Ω–æ –ø–æ–ª—É—á–∏—Ç —Å–≤–æ—é –ø—è—Ç–µ—Ä–∫—É, —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω–∞ –ª–∏ –æ–Ω–∞, –µ—Å–ª–∏ —É —Ç–µ–±—è –Ω–∞ —Ä—É–∫–∞—Ö –º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω—ã–π —Å–µ—Ä—Ç–∏—Ñ–∏–∫–∞—Ç –∏ –ø—Ä–µ–∫—Ä–∞—Å–Ω—ã–π –∞–Ω–≥–ª–∏–π—Å–∫–∏–π –ß—Ç–æ —Ç–∞–∫–æ–µ –º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω—ã–π —ç–∫–∑–∞–º–µ–Ω? –≠—Ç–æ —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ —Ä–∞–∑—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã–π —Ç–µ—Å—Ç, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –≤–ª–∞–¥–µ–Ω–∏—è —è–∑—ã–∫–æ–º. –ü–æ–ª—É—á–µ–Ω–Ω—ã–π —Å–µ—Ä—Ç–∏—Ñ–∏–∫–∞—Ç –ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è –≤ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–µ —Å—Ç—Ä–∞–Ω –∏ –º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω—ã—Ö –∫–æ–º–ø–∞–Ω–∏–π. –°–∞–º —Ç–µ—Å—Ç –ø—Ä–æ–≤–µ—Ä—è–µ—Ç –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ö–æ—Ä–æ—à–æ —É—á–∞—â–∏–π—Å—è –≤–ª–∞–¥–µ–µ—Ç —É—Å—Ç–Ω–æ–π –∏ –ø–∏—Å—å–º–µ–Ω–Ω–æ–π —Ä–µ—á—å—é, –ø–æ–Ω–∏–º–∞–µ—Ç —Ä–µ—á—å –Ω–∞ —Å–ª—É—Ö, –≤–ª–∞–¥–µ–µ—Ç –ª–µ–∫—Å–∏—á–µ—Å–∫–∏–º –∏ –≥—Ä–∞–º–º–∞—Ç–∏—á–µ—Å–∫–∏–º –º–∞—Ç–µ—Ä–∏–∞–ª–æ–º –Ω–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º —É—Ä–æ–≤–Ω–µ. –ï—Å—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –º–ª–∞–¥—à–∏—Ö —à–∫–æ–ª—å–Ω–∏–∫–æ–≤, –µ—Å—Ç—å –¥–ª—è –ø–æ–¥—Ä–æ—Å—Ç–∫–æ–≤ –∏ –≤–∑—Ä–æ—Å–ª—ã—Ö. –ö–æ–Ω–µ—á–Ω–æ –∫ –Ω–µ–º—É –Ω–µ–æ–±—Ö–æ–¥–∏–º–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–∞—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞, –Ω–æ –ø–æ—Å–ª–µ –ø–æ–ª—É—á–µ–Ω–∏—è —Å–µ—Ä—Ç–∏—Ñ–∏–∫–∞—Ç–∞, —É—á–µ–Ω–∏–∫ –ø–æ–ª—É—á–∞–µ—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—É—é –º–æ—Ç–∏–≤–∞—Ü–∏—é –ø—Ä–æ–¥–æ–ª–∂–∞—Ç—å –æ–±—É—á–µ–Ω–∏–µ, –∞ —Ä–æ–¥–∏—Ç–µ–ª—å –≤–∏–¥–∏—Ç —Ä–µ–∞–ª—å–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç! –ò–∑—É—á–∞–π—Ç–µ —è–∑—ã–∫–∏ —Ç–∞–∫, —á—Ç–æ–±—ã –Ω—Ä–∞–≤–∏–ª–æ—Å—å! –Ø–∑—ã–∫–æ–≤–æ–π —Ü–µ–Ω—Ç—Ä Sherwood –Ω–∞ –¢–∏—Ç–æ–≤–∞, 242/2 –ù–∞–ø–∏—à–∏—Ç–µ –Ω–∞–º –∏ —É–∑–Ω–∞–π—Ç–µ –≤—Å–µ –ø–æ–¥—Ä–æ–±–Ω–æ - https://api.whatsapp.com/send/?phone=79628389934 –ü–æ—Å—Ç 3 –ï—â–µ –ø—è—Ç—å –∫–æ–ø–µ–µ–∫. –í–æ—Ç –º–æ–∂–Ω–æ –º–Ω–æ–≥–æ –≥–æ–≤–æ—Ä–∏—Ç—å, —á—Ç–æ –ø–µ—Ä–µ–¥ —à–∫–æ–ª–æ–π –Ω—É–∂–Ω–æ –Ω–∞—É—á–∏—Ç—å—Å—è —Å—á–∏—Ç–∞—Ç—å, –ø–∏—Å–∞—Ç—å –∏ –ø—Ä–æ—á–µ–µ. –î–∞-–¥–∞-–¥–∞, —ç—Ç–æ –≤–∞–∂–Ω–æ. –ù–æ, –∑–Ω–∞–µ—Ç–µ, —á—Ç–æ –º—ã —Å–¥–µ–ª–∞–ª–∏ –∫–ª—é—á–µ–≤—ã–º –≤ –Ω–∞—à–µ–π –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏? –í–æ—Ç –ø—Ä—è–º –∫—Ä—É–ø–Ω—ã–º–∏ –±—É–∫–≤–∞–º–∏ –Ω–∞–ø–∏—à–µ–º, –Ω–µ –ø—É–≥–∞–π—Ç–µ—Å—å. –ù–ê–£–ß–ò–ú –î–ï–ô–°–¢–í–û–í–ê–¢–¨ –í –ö–û–ú–ê–ù–î–ï –ò –û–†–ò–ï–ù–¢–ò–†–û–í–ê–¢–¨–°–Ø –í –ö–û–õ–õ–ï–ö–¢–ò–í–ï –£–∑–Ω–∞–π—Ç–µ –±–æ–ª—å—à–µ, –∑–∞–π–¥—è –∫ –Ω–∞–º –≤ —Ü–µ–Ω—Ç—Ä –∏–ª–∏ –Ω–∞–ø–∏—à–∏—Ç–µ –≤ –î–∏—Ä–µ–∫—Ç –ò–ª–∏ –ø–æ—á–∏—Ç–∞–π—Ç–µ –Ω–∞ —Å—Ç—Ä–∞–Ω–∏—Ü–µ - https://www.sherwood-school.ru/podgotovka_k_shkole –í–æ—Ç –∏–º–µ–Ω–Ω–æ —Ç–∞–∫, –∏ —ç—Ç–æ –±—ã–ª–∏ 5 –∫–æ–ø–µ–µ–∫ –∏ –ø–æ—ç—Ç–æ–º—É –Ω–µ –±—É–¥–µ–º —Ä–∞—Å–ø–∏—Å—ã–≤–∞—Ç—å —Ü–µ–ª–æ–µ –ø–æ–ª–æ—Ç–Ω–æ. –•–æ—Ä–æ—à–µ–≥–æ –¥–Ω—è! –° –ª—é–±–æ–≤—å—é, Sherwood! –£–∑–Ω–∞–π—Ç–µ –±–æ–ª—å—à–µ –æ –Ω–∞—à–∏—Ö –ø—Ä–æ–≥—Ä–∞–º–º–∞—Ö, –Ω–∞–ø–∏—Å–∞–≤ –Ω–∞–º –≤ WhatsApp - https://api.whatsapp.com/send/?phone=79628389934 

–ü–æ—Å—Ç 3 
–í–æ–æ–±—â–µ –º–æ–∂–Ω–æ –±—ã–ª–æ –±—ã –º–Ω–æ–≥–æ–µ —Å–∫–∞–∑–∞—Ç—å, –Ω–æ —Å–∫–∞–∂–µ–º —Å–∞–º—É—é –ø–µ—Ä–≤—É—é –º—ã—Å–ª—å, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–Ω–∏–∫–ª–∞ —É –Ω–∞—Å, –∫–æ–≥–¥–∞ –º—ã –ø—Ä–æ—á–∏—Ç–∞–ª–∏ —ç—Ç–∏ –æ—Ç–∑—ã–≤—ã. –†–∞–∑–º–µ—Å—Ç–∏–º –ª–∏—à—å –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ. –°–ø–∞—Å–∏–±–æ –≤–∞–º –∑–∞ —Ç–æ, —á—Ç–æ –¥–æ–≤–µ—Ä—è–µ—Ç–µ –Ω–∞–º —Å–∞–º–æ–µ —Ü–µ–Ω–Ω–æ–µ ‚Äî —É—á–∞—Å—Ç–∏–µ –≤ –≤–æ—Å–ø–∏—Ç–∞–Ω–∏–∏ –∏ –≤–∑—Ä–æ—Å–ª–µ–Ω–∏–∏ –¥–µ—Ç–µ–π. –ú—ã —á—É–≤—Å—Ç–≤—É–µ–º –±–æ–ª—å—à—É—é –æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ—Å—Ç—å –ø–µ—Ä–µ–¥ –≤–∞–º–∏! –û—á–µ–Ω—å –±–ª–∞–≥–æ–¥–∞—Ä–Ω—ã –≤—Å–µ–º —Ä–æ–¥–∏—Ç–µ–ª—è–º –∑–∞ —Ç–µ–ø–ª—ã–µ —Å–ª–æ–≤–∞, –∑–∞ —Ç–æ, —á—Ç–æ –æ–Ω–∏ –ø–æ–º–æ–≥–∞—é—Ç –Ω–∞–º –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ –æ–±—É—á–µ–Ω–∏—è. –°–ø–∞—Å–∏–±–æ –≤–∞–º –≤—Å–µ–º! –ë—É–¥–µ–º –∏ –¥–∞–ª—å—à–µ –∏–∑—É—á–∞—Ç—å —è–∑—ã–∫–∏ —Ç–∞–∫, —á—Ç–æ–±—ã –Ω—Ä–∞–≤–∏–ª–æ—Å—å! –ü–æ—Å—Ç 2 –¢–∞–∫, –¥–∞–≤–∞–π—Ç–µ –ø—Ä–æ–≤–µ—Ä–∏–º, –≤—Å–µ –ª–∏ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–∞ –æ–±—É—á–µ–Ω–∏—è –≤ —è–∑—ã–∫–æ–≤–æ–º —Ü–µ–Ω—Ç—Ä–µ —Å –õ–∏—Ü–µ–Ω–∑–∏–µ–π –≤—ã –∏—Å–ø–æ–ª—å–∑—É–µ—Ç–µ? –í 2021 –≥–æ–¥—É –º—ã —Å–¥–µ–ª–∞–ª–∏ —ç—Ç–æ—Ç –Ω–µ–ø—Ä–æ—Å—Ç–æ–π —à–∞–≥ –∏ –ø–æ–ª—É—á–∏–ª–∏ –ª–∏—Ü–µ–Ω–∑–∏—é –Ω–∞ –æ–±—Ä–∞–∑–æ–≤–∞—Ç–µ–ª—å–Ω—É—é –¥–µ—è—Ç–µ–ª—å–Ω–æ—Å—Ç—å. –ù–∞–º –Ω—É–∂–Ω–æ –±—ã–ª–æ –ø–æ–ª—É—á–∏—Ç—å —Ä–∞–∑—Ä–µ—à–µ–Ω–∏–µ –æ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –æ—á–µ–Ω—å —Å–µ—Ä—å–µ–∑–Ω—ã—Ö –∏–Ω—Å—Ç–∞–Ω—Ü–∏–π - –¶–µ–Ω—Ç—Ä–∞ –≥–∏–≥–∏–µ–Ω—ã –∏ —ç–ø–∏–¥–µ–º–∏–æ–ª–æ–≥–∏–∏, –†–æ—Å–ø–æ—Ç—Ä–µ–±–Ω–∞–¥–∑–æ—Ä–∞, –ú–∏–Ω–∏—Å—Ç–µ—Ä—Å—Ç–≤–∞ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è, –ø–ª—é—Å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –≤—Å–µ–º –Ω–æ—Ä–º–∞–º –ú–ß–°. –ù–∞—Å –ø—Ä–æ–≤–µ—Ä—è–ª–∏, –∫ –Ω–∞–º –≤—ã–µ–∑–∂–∞–ª–∏ —Å–ø–µ—Ü–∏–∞–ª–∏—Å—Ç—ã, –æ—Å–º–∞—Ç—Ä–∏–≤–∞–ª–∏ –≤—Å–µ –Ω–∞ –º–µ—Å—Ç–µ, —Å–º–æ—Ç—Ä–µ–ª–∏ –≤—Å–µ –±—É–º–∞–≥–∏, –∏ –∑–∞–ø—Ä–∞—à–∏–≤–∞–ª–∏ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏—è –≤ —Å–º–µ–∂–Ω—ã—Ö –≤–µ–¥–æ–º–æ—Å—Ç—è—Ö. –í—ã –º–æ–∂–µ—Ç–µ –±—ã—Ç—å —É–≤–µ—Ä–µ–Ω—ã, —á—Ç–æ –Ω–∞—Å —Ü–µ–Ω—Ç—Ä —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –≤—Å–µ–º —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º: ‚ÅÉ —Å–∞–º–æ –ø–æ–º–µ—â–µ–Ω–∏–µ ‚ÅÉ –æ—Å–≤–µ—â–µ–Ω–∏–µ –∫–∞–±–∏–Ω–µ—Ç–æ–≤ ‚ÅÉ —Å–∞–Ω–∏—Ç–∞—Ä–Ω—ã–µ –Ω–æ—Ä–º—ã ‚ÅÉ –æ—Å–Ω–∞—â–µ–Ω–∏–µ –∫–∞–±–∏–Ω–µ—Ç–æ–≤ ‚ÅÉ –º–µ–±–µ–ª—å ‚ÅÉ –ø–æ–∂–∞—Ä–Ω–∞—è –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç—å ‚ÅÉ –ø–µ–¥–∞–≥–æ–≥–∏—á–µ—Å–∫–∏–π —Å–æ—Å—Ç–∞–≤ ‚ÅÉ —É—á–µ–±–Ω—ã–µ –ø—Ä–æ–≥—Ä–∞–º–º—ã –ò —Å–∞–º–æ–µ –≥–ª–∞–≤–Ω–æ–µ ‚Äî –≤—ã –º–æ–∂–µ—Ç–µ –æ—Ñ–æ—Ä–º–∏—Ç—å –Ω–∞–ª–æ–≥–æ–≤—ã–π –≤—ã—á–µ—Ç –∏ –ø–æ–ª—É—á–∏—Ç—å 13% –æ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –æ–±—É—á–µ–Ω–∏—è . –ê —Ç–∞–∫–∂–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–∞—Ç–µ—Ä–∏–Ω—Å–∫–∏–π –∫–∞–ø–∏—Ç–∞–ª –¥–ª—è –æ–ø–ª–∞—Ç—ã –æ–±—É—á–µ–Ω–∏—è! –ü–æ—Å—Ç 4 
–ó–Ω–∞–µ—Ç–µ, —á—Ç–æ —Å–∞–º–æ–µ —Å—Ç—Ä–∞—à–Ω–æ–µ –¥–ª—è –Ω–∞—Å –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —É—Ä–æ–∫–∞? –í–æ—Ç –ø—Ä—è–º —Ç–∞–∫ —Å—Ç—Ä–∞—à–Ω–æ, —á—Ç–æ –µ–¥–∏–º –ø–ª–æ—Ö–æ. –í–æ—Ç –ø—Ä–∞–≤–¥–∞! –°–∞–º–æ–µ —Å—Ç—Ä–∞—à–Ω–æ–µ, –µ—Å–ª–∏ –º—ã –Ω–µ —Å–º–æ–∂–µ–º —Å–¥–µ–ª–∞—Ç—å —Ç–∞–∫, —á—Ç–æ–±—ã –Ω–∞ —É—Ä–æ–∫–∞—Ö –¥–µ—Ç—è–º –±—ã–ª–æ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ –∏ –Ω–µ —Å–∫—É—á–Ω–æ. –ò–Ω—Ç–µ—Ä–µ—Å –∫ –æ–±—É—á–µ–Ω–∏—é –∏–≥—Ä–∞–µ—Ç –æ–≥—Ä–æ–º–Ω—É—é —Ä–æ–ª—å –≤ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–∏ —É—Å–ø–µ—à–Ω–æ–≥–æ –æ–±—Ä–∞–∑–æ–≤–∞—Ç–µ–ª—å–Ω–æ–≥–æ –ø—É—Ç–∏ –∏, —á—Ç–æ –µ—â–µ –≤–∞–∂–Ω–µ–µ, —Å—á–∞—Å—Ç–ª–∏–≤–æ–π –∂–∏–∑–Ω–∏. –ò –µ—Å–ª–∏ —ç—Ç–æ–≥–æ –Ω–µ —Å–¥–µ–ª–∞—Ç—å, —Ç–æ: –ë—É–¥–µ—Ç –ø–æ—Ç–µ—Ä—è –º–æ—Ç–∏–≤–∞—Ü–∏–∏ –ë—É–¥–µ—Ç —Å—Ç—Ä–µ—Å—Å –∏ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–µ —ç–º–æ—Ü–∏–∏ –ë—É–¥—É—Ç –ø–ª–æ—Ö–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ë—É–¥–µ—Ç –ø–æ—Ç–µ—Ä—è —Ä–∞–¥–æ—Å—Ç–∏ –æ—Ç –æ–±—É—á–µ–Ω–∏—è –ü–æ—ç—Ç–æ–º—É –º—ã –µ–∂–µ–¥–Ω–µ–≤–Ω–æ —Å–æ–∑–¥–∞–µ–º —É—Å–ª–æ–≤–∏—è, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –¥–µ—Ç–∏ –ø–æ–ª—É—á–∞—é—Ç —É–¥–æ–≤–æ–ª—å—Å—Ç–≤–∏–µ –æ—Ç –æ–±—É—á–µ–Ω–∏—è. –ü–æ–æ—â—Ä—è–µ–º –∏—Ö –∏–Ω—Ç–µ—Ä–µ—Å—ã, –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º –∏—Ö —Å—Ç—Ä–µ–º–ª–µ–Ω–∏—è, –º–æ—Ç–∏–≤–∏—Ä—É–µ–º. –¢–æ–ª—å–∫–æ —Ç–∞–∫ —Ä–µ–±–µ–Ω–æ–∫ —Å–º–æ–∂–µ—Ç —Ä–∞—Å–∫—Ä—ã—Ç—å —Å–≤–æ–π –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å —É—Å–ø–µ—à–Ω–æ–µ –±—É–¥—É—â–µ–µ, –Ω–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ–µ —Ä–∞–¥–æ—Å—Ç—å—é –∏ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–∏–µ–º.

–ü–æ—Å—Ç 5
–¢—Ä–µ–≤–æ–≥—É –Ω–∏–∫—Ç–æ –Ω–∏–∫–æ–≥–¥–∞ –Ω–µ –∂–¥–µ—Ç, –∞ –æ–Ω–∞ –≤—Å–µ —Ä–∞–≤–Ω–æ –ø—Ä–∏—Ö–æ–¥–∏—Ç
–°–µ—Ä–¥—Ü–µ –¥–∏–∫–æ —Å—Ç—É—á–∏—Ç, –º—ã—Å–ª–∏ –º–µ—á—É—Ç—Å—è, –∞ –≤ –≥–æ—Ä–ª–µ –∑–∞—Å—Ç—Ä–µ–≤–∞–µ—Ç –∫–æ–º–æ–∫. –ê –≥–ª–∞–≤–Ω–æ–µ, –∫–∞–∫ —É—Å–ø–æ–∫–æ–∏—Ç—å—Å—è ‚Äî –Ω–µ–ø–æ–Ω—è—Ç–Ω–æ.
–ò–Ω—Ç—Ä–æ–≤–µ—Ä—Ç —Å–ø–µ—à–∏—Ç –Ω–∞ –ø–æ–º–æ—â—å! –ú—ã –∑–∞–ø–∏—Å–∞–ª–∏ –ª–µ–∫—Ü–∏—é ¬´–ö–∞–∫ —É–ø—Ä–∞–≤–ª—è—Ç—å —Ç—Ä–µ–≤–æ–≥–æ–π: –∑–∞–º–µ—á–∞—Ç—å, –ø–æ–Ω–∏–º–∞—Ç—å, –æ—Ç–ø—É—Å–∫–∞—Ç—å¬ª, –∏ –≤ –Ω–µ–π —Ä–∞—Å—Å–∫–∞–∑—ã–≤–∞–µ–º:

üîπ –ö–∞–∫ –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —Ç—Ä–µ–≤–æ–≥—É –¥–æ —Ç–æ–≥–æ, –∫–∞–∫ –æ–Ω–∞ –≤—ã–π–¥–µ—Ç –∏–∑-–ø–æ–¥ –∫–æ–Ω—Ç—Ä–æ–ª—è
üîπ –ß—Ç–æ –æ–Ω–∞ –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ —Ö–æ—á–µ—Ç —Ç–µ–±–µ —Å–∫–∞–∑–∞—Ç—å 
üîπ –ö–∞–∫–∏–µ —à–∞–≥–∏ —Å–¥–µ–ª–∞—Ç—å, —á—Ç–æ —É—Å–ø–æ–∫–æ–∏—Ç—å—Å—è, –∫–æ–≥–¥–∞ —É–∂–µ –≤—Å—Ç—Ä–µ–≤–æ–∂–∏–ª—Å—è

–ò —Å–∞–º–æ–µ –ø—Ä–∏—è—Ç–Ω–æ–µ ‚Äî –º—ã –æ—Ç–∫—Ä—ã–ª–∏ –±–µ—Å–ø–ª–∞—Ç–Ω—ã–π –¥–æ—Å—Ç—É–ø –Ω–∞ 48 —á–∞—Å–æ–≤, —á—Ç–æ–±—ã —ç—Ç–∏ –∑–Ω–∞–Ω–∏—è –±—ã–ª–∏ —É –≤—Å–µ—Ö. –°–∫–∞—á–∏–≤–∞–π –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ
P.S.: –ò –Ω–∞–ø–∏—à–∏, –∫–∞–∫–∏–µ –µ—â–µ –Ω–∞—à–∏ –ª–µ–∫—Ü–∏–∏ –ø–æ–º–æ–≥–ª–∏ —Ç–µ–±–µ –≤ –±–æ—Ä—å–±–µ —Å –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–º–∏ –¥–µ–º–æ–Ω–∞–º–∏, –æ–±–º–µ–Ω—è–µ–º—Å—è –Ω–∞—Ö–æ–¥–∫–∞–º–∏!

–ü–æ—Å—Ç 6
–ö–∞–∫ —Ä–∞–∑–æ–±—Ä–∞—Ç—å—Å—è –≤ –∏—Å–∫—É—Å—Å—Ç–≤–µ? –ò –ø–æ—á–µ–º—É —Ñ–∏–ª–æ—Å–æ—Ñ—Å—Ç–≤–æ–≤–∞—Ç—å ‚Äî –æ—á–µ–Ω—å –≤–∞–∂–Ω—ã–π –Ω–∞–≤—ã–∫?

–û–± —ç—Ç–æ–º, –∞ –µ—â–µ –æ —Ç—Ä—É–¥–Ω–æ—Å—Ç—è—Ö –ø—Ä–æ—Ñ–µ—Å—Å–∏–∏ –ø—Å–∏—Ö–æ–ª–æ–≥–∞, –≤–Ω—É—Ç—Ä—è–Ω–∫–µ –∫–∏–Ω–æ–∫—Ä–∏—Ç–∏–∫–∏ –∏ –º–Ω–æ–≥–æ–º –¥—Ä—É–≥–æ–º –º—ã —Ä–∞–∑–≥–æ–≤–∞—Ä–∏–≤–∞–µ–º –Ω–∞ –±–µ—Å–ø–ª–∞—Ç–Ω—ã—Ö –≤–µ–±–∏–Ω–∞—Ä–∞—Ö –ò–Ω—Ç—Ä–æ–≤–µ—Ä—Ç–∞!

–¢—Ä–∏ —Ä–∞–∑–∞ –≤ –Ω–µ–¥–µ–ª—é –≤ 19:00 –Ω–∞—à–∏ –ª–µ–∫—Ç–æ—Ä—ã –≤—ã—Ö–æ–¥—è—Ç –≤ –ø—Ä—è–º–æ–π —ç—Ñ–∏—Ä, —Ä–∞—Å—Å–∫–∞–∑—ã–≤–∞—é—Ç –æ —Å–≤–æ–µ–º –¥–µ–ª–µ –∏–∑–Ω—É—Ç—Ä–∏ –∏ –æ—Ç–≤–µ—á–∞—é—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã. –ó–≤—É—á–∏—Ç –∫–∞–∫ –æ—Ç–ª–∏—á–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç –¥–ª—è –≤–µ—á–µ—Ä–∞ –≤ —Ö–æ—Ä–æ—à–µ–π –∫–æ–º–ø–∞–Ω–∏–∏, –¥–∞?

–ê –≥–ª–∞–≤–Ω–æ–µ ‚Äî –ø–æ–ø–∞—Å—Ç—å –Ω–∞ —ç—Ç–∏ –≤–µ–±–∏–Ω–∞—Ä—ã —Å—É–ø–µ—Ä-–ø—Ä–æ—Å—Ç–æ. –í–æ—Ç 3 —à–∞–≥–∞:
1. –ó–∞—Ö–æ–¥–∏—à—å –Ω–∞ —ç—Ç—É —Å—Ç—Ä–∞–Ω–∏—á–∫—É —Å —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–µ–º (https://online.artforintrovert.ru/timetable?utm_source=tg&utm_medium=social&utm_campaign=timetable&utm_content=post&utm_term=180925)
2. –í—ã–±–∏—Ä–∞–µ—à—å —Å–∞–º—ã–π –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–π –≤–µ–± (–∏–ª–∏ –≤—Å–µ —Å—Ä–∞–∑—É) –∏ –∑–∞–ø–∏—Å—ã–≤–∞–µ—à—å—Å—è –Ω–∞ –Ω–µ–≥–æ
3. –ü–æ–¥–∫–ª—é—á–∞–µ—à—å—Å—è –≤ –Ω–∞–∑–Ω–∞—á–µ–Ω–Ω—ã–π –¥–µ–Ω—å –≤ 19:00 –∏ –∫–∞–π—Ñ—É–µ—à—å

–í–µ–±–∏–Ω–∞—Ä—ã –∏–¥—É—Ç –ø—Ä–∏–º–µ—Ä–Ω–æ —á–∞—Å –∏–ª–∏ –Ω–µ–º–Ω–æ–≥–æ –±–æ–ª—å—à–µ, –∏ –∫–∞–∂–¥—ã–π –∏–∑ –Ω–∏—Ö –Ω–µ–ø–æ–≤—Ç–æ—Ä–∏–º, –≤–µ–¥—å –º—ã –∏—Ö –Ω–µ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º. 
–ï—Å–ª–∏ –æ—Å—Ç–∞–ª–∏—Å—å –≤–æ–ø—Ä–æ—Å—ã, —Å–ø—Ä–∞—à–∏–≤–∞–π –≤ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è—Ö, –∞ —Ç–∞–∫, –Ω–∞–¥–µ—é—Å—å, –≤—Å—Ç—Ä–µ—Ç–∏–º—Å—è —Å —Ç–æ–±–æ–π –Ω–∞ –≤–µ–±–∞—Ö üòâ

–ü–æ—Å—Ç 7

–ë—ã–≤–∞–µ—Ç, –ø—Ä–æ—Å—ã–ø–∞–µ—à—å—Å—è –≤ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫ –∏ —á—É–≤—Å—Ç–≤—É–µ—à—å: ¬´–•–æ—á—É –Ω–∞—á–∞—Ç—å —ç—Ç—É –Ω–µ–¥–µ–ª—é –ø–æ-–¥—Ä—É–≥–æ–º—É‚Ä¶ –ø–æ-–∂–∏–≤–æ–º—É, —Å –∑–∞–±–æ—Ç–æ–π –æ —Å–µ–±–µ, —Å –º–∞–ª–µ–Ω—å–∫–∏–º, –Ω–æ –≤–∞–∂–Ω—ã–º —à–∞–≥–æ–º –≤–ø–µ—Ä–µ–¥¬ª ‚ò∫Ô∏è
–ò –≤ —Ç–∞–∫–∏–µ –º–æ–º–µ–Ω—Ç—ã –æ—Å–æ–±–µ–Ω–Ω–æ –≤–∞–∂–Ω–æ –∏–º–µ—Ç—å –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ, –≥–¥–µ –º–æ–∂–Ω–æ –≤–¥–æ—Ö–Ω–æ–≤–∏—Ç—å—Å—è, –ø—Ä–æ–∫–∞—á–∞—Ç—å –Ω–∞–≤—ã–∫–∏, –Ω–∞–π—Ç–∏ –æ—Ç–≤–µ—Ç—ã –∏–ª–∏ –ø—Ä–æ—Å—Ç–æ –ø–æ—á—É–≤—Å—Ç–≤–æ–≤–∞—Ç—å –ø–æ–¥–¥–µ—Ä–∂–∫—É.

–ù–∞ —ç—Ç–æ–π –Ω–µ–¥–µ–ª–µ –≤ SPACE ‚Äî –∫–∞–∫ —Ä–∞–∑ —Ç–æ —Å–∞–º–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ.
–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ, –≥–¥–µ –∫–∞–∂–¥–∞—è –≤—Å—Ç—Ä–µ—á–∞ –ø–æ–º–æ–≥–∞–µ—Ç —Å—Ç–∞—Ç—å –∫ —Å–µ–±–µ —á—É—Ç–æ—á–∫—É –±–ª–∏–∂–µ, —Å–º–µ–ª–µ–µ, —É–≤–µ—Ä–µ–Ω–Ω–µ–µ.

–ò –¥–∞–≤–∞–π –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –∂–µ —Ç–µ–±–µ –¥–æ—Å—Ç—É–ø–Ω–æ –≤ –±–ª–∏–∂–∞–π—à–∏–µ –¥–Ω–∏ üî•

–ß—Ç–æ –∂–¥—ë—Ç —Ç–µ–±—è –Ω–∞ —ç—Ç–æ–π –Ω–µ–¥–µ–ª–µ

–° 27.10 –ø–æ 02.11 —Ç—ã –º–æ–∂–µ—à—å –ø–æ—Å–µ—Ç–∏—Ç—å 87 –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π:
‚Äî 72 –æ–Ω–ª–∞–π–Ω ‚Äî –≤–∫–ª—é—á–∞–π—Å—è –æ—Ç–∫—É–¥–∞ —É–¥–æ–±–Ω–æ;
‚Äî 15 –æ—Ñ—Ñ–ª–∞–π–Ω ‚Äî –µ—Å–ª–∏ —Ö–æ—á–µ—Ç—Å—è –∂–∏–≤–æ–≥–æ –∫–æ–Ω—Ç–∞–∫—Ç–∞;

–ê –µ—Å–ª–∏ —É —Ç–µ–±—è –µ—Å—Ç—å SPACE Pass, —Ç–æ –≤–æ–æ–±—â–µ –∫–ª–∞—Å—Å:
‚Äî 17 –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π –æ–Ω–ª–∞–π–Ω
‚Äî 3 –æ—Ñ—Ñ–ª–∞–π–Ω
–¥–æ—Å—Ç—É–ø–Ω—ã —Ç–µ–±–µ –±–µ–∑ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π. –≠—Ç–æ –æ—Ç–ª–∏—á–Ω—ã–π —Å–ø–æ—Å–æ–± –≥–ª—É–±–∂–µ –ø–æ–≥—Ä—É–∑–∏—Ç—å—Å—è –≤ —Ç–µ–º—ã, –∫–æ—Ç–æ—Ä—ã–µ –¥–∞–≤–Ω–æ —á—É–≤—Å—Ç–≤—É—é—Ç—Å—è –≤–∞–∂–Ω—ã–º–∏ üôè

–ò —á—Ç–æ–±—ã —Ç—ã –ø–æ–Ω–∏–º–∞–ª –º–∞—Å—à—Ç–∞–±: –∑–∞ –≤—Å—ë –≤—Ä–µ–º—è –≤ SPACE —É–∂–µ –ø—Ä–æ—à–ª–æ 7438 –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π ‚Äî —Ç—ã—Å—è—á–∏ –∏—Å—Ç–æ—Ä–∏–π, —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏–π –∏ —Ç–æ—á–µ–∫ –æ–ø–æ—Ä—ã.

–î–∞–≤–∞–π –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ —É –Ω–∞—Å –±—É–¥–µ—Ç –ø–æ –¥–Ω—è–º –Ω–µ–¥–µ–ª–∏

–ù–∏–∂–µ ‚Äî –∫–æ—Ä–æ—Ç–∫–∏–µ –æ–±–∑–æ—Ä –∫–∞–∂–¥–æ–≥–æ –¥–Ω—è.

–ò–ò SPACE –≤—ã–±—Ä–∞–ª —Ç—Ä–∏ —Å–ª—É—á–∞–π–Ω—ã—Ö –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è –Ω–∞ –¥–µ–Ω—å, —á—Ç–æ–±—ã —Ç—ã –±—ã—Å—Ç—Ä–æ –Ω–∞—à—ë–ª —Ç–æ, —á—Ç–æ –æ—Ç–∫–ª–∏–∫–∞–µ—Ç—Å—è. –û—Å—Ç–∞–ª—å–Ω–æ–µ ‚Äî –≤ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–∏.
–ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫ ‚Äî –º—è–≥–∫–∞—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –∏ –≤–Ω—É—Ç—Ä–µ–Ω–Ω—è—è –æ–ø–æ—Ä–∞

–í —ç—Ç–æ—Ç –¥–µ–Ω—å —Ç—ã –º–æ–∂–µ—à—å –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —ç–Ω–µ—Ä–≥–∏—é, —É—Å–ª—ã—à–∞—Ç—å —Ç–µ–ª–æ –∏ –≤–µ—Ä–Ω—É—Ç—å —Ñ–æ–∫—É—Å. –≠—Ç–æ –ø–æ–º–æ–∂–µ—Ç –≤–æ–π—Ç–∏ –≤ –Ω–µ–¥–µ–ª—é —Å–ø–æ–∫–æ–π–Ω–æ –∏ –æ—Å–æ–∑–Ω–∞–Ω–Ω–æ.

‚Ä¢ –î–ï–¢–û–ö–° —Å–æ –°–º—ã—Å–ª–æ–º ‚Äî https://app.elsspace.ru/activity/34528
‚Ä¢ –ö–æ–Ω—Å—É–ª—å—Ç–∞—Ü–∏—è ‚Äî https://app.elsspace.ru/activity/38213
‚Ä¢ –î–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∞ —Å–æ–º–∞—Ç–∏–ø–æ–ª–æ–≥–∞ ‚Äî https://app.elsspace.ru/activity/27813

–í—Ç–æ—Ä–Ω–∏–∫ ‚Äî —è—Å–Ω–æ—Å—Ç—å, —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –∏ —Ä–µ—à–µ–Ω–∏—è

–¢—ã —É–∑–Ω–∞–µ—à—å, –∫–∞–∫ —Ä–∞–±–æ—Ç–∞—Ç—å —Å —Å–æ—Å—Ç–æ—è–Ω–∏—è–º–∏ –∏ –≤–Ω–µ—à–Ω–∏–º –æ–±—Ä–∞–∑–æ–º —Ç–∞–∫, —á—Ç–æ–±—ã —á—É–≤—Å—Ç–≤–æ–≤–∞—Ç—å —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç—å –∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –≤ –¥–µ–ª–∞—Ö –∏ –æ–±—â–µ–Ω–∏–∏.

‚Ä¢ –ö–æ—É—á-—Å–µ—Å—Å–∏—è ‚Äî https://app.elsspace.ru/activity/38293
‚Ä¢ –ö–æ–Ω—Å—É–ª—å—Ç–∞—Ü–∏—è –ø–æ –ø—Å–∏—Ö–æ—Å–æ–º–∞—Ç–∏–∫–µ ‚Äî https://app.elsspace.ru/activity/38428
‚Ä¢ –ö–æ–Ω—Å—É–ª—å—Ç–∞—Ü–∏—è –ø–æ –§–æ—Ä–º—É–ª–µ —Å—Ç–∏–ª—è ‚Äî https://app.elsspace.ru/activity/38108

–°—Ä–µ–¥–∞ ‚Äî —Ä–∞–∑–≤–∏—Ç–∏–µ –Ω–∞–≤—ã–∫–æ–≤ –∏ –Ω–æ–≤–æ–µ –≤–∏–¥–µ–Ω–∏–µ

–í —ç—Ç–æ—Ç –¥–µ–Ω—å —Ç—ã –º–æ–∂–µ—à—å —Ä–∞—Å—à–∏—Ä–∏—Ç—å —Å–≤–æ—é –∫–∞—Ä—Ç–∏–Ω—É –º–∏—Ä–∞ –∏ –ø–æ–º–æ—á—å –¥–µ—Ç—è–º —Ä–∞–∑–≤–∏–≤–∞—Ç—å—Å—è —á–µ—Ä–µ–∑ –∏–Ω—Ç–µ–ª–ª–µ–∫—Ç –∏ —Ç–≤–æ—Ä—á–µ—Å—Ç–≤–æ.

‚Ä¢ –®–∞—Ö–º–∞—Ç–Ω—ã–π VIP-–∫—Ä—É–∂–æ–∫ ‚Äî https://app.elsspace.ru/activity/38448
‚Ä¢ –ö–æ–Ω—Å—É–ª—å—Ç–∞—Ü–∏—è ¬´–§–æ—Ä–º—É–ª–∞ —Å—Ç–∏–ª—è¬ª ‚Äî https://app.elsspace.ru/activity/38538
‚Ä¢ –ò–∑–æ–±–∏–ª–∏–µ –∏ —Ä–µ–≥—Ä–µ—Å—Å ‚Äî https://app.elsspace.ru/activity/38503

–ß–µ—Ç–≤–µ—Ä–≥ ‚Äî –¥–≤–∏–∂–µ–Ω–∏–µ, —Ä–µ—Å—É—Ä—Å –∏ –≤–¥–æ—Ö–Ω–æ–≤–µ–Ω–∏–µ

–¢—ã –º–æ–∂–µ—à—å –≤—Å—Ç—Ä—è—Ö–Ω—É—Ç—å —Ç–µ–ª–æ, –≤–¥–æ—Ö–Ω–æ–≤–∏—Ç—å—Å—è –ø—É—Ç–µ—à–µ—Å—Ç–≤–∏—è–º–∏ –∏ –ø–æ–ª–Ω–æ—Å—Ç—å—é –ø–µ—Ä–µ–∑–∞–≥—Ä—É–∑–∏—Ç—å—Å—è.

‚Ä¢ –ú–∞—Ä–∞—Ñ–æ–Ω –®–∏–∫–∞—Ä–Ω–∞—è –ø–æ—Ö–æ–¥–∫–∞ 2.0 ‚Äî https://app.elsspace.ru/activity/38643
‚Ä¢ –ê–≤—Ç–æ—Ä—Å–∫–∏–π —Ç—É—Ä –≤ –ò–æ—Ä–¥–∞–Ω–∏—é ‚Äî https://app.elsspace.ru/activity/26379
‚Ä¢ –†–ï–ó–ï–¢ ‚Äî –ø–µ—Ä–µ–∑–∞–≥—Ä—É–∑–∫–∞ –∏ —Ä–∞—Å—Å–ª–∞–±–ª–µ–Ω–∏–µ ‚Äî https://app.elsspace.ru/activity/38673

–ü—è—Ç–Ω–∏—Ü–∞ ‚Äî –æ—Å–æ–∑–Ω–∞–Ω–Ω–æ—Å—Ç—å –∏ –∑–∞–±–æ—Ç–∞ –æ —Å–µ–±–µ

–≠—Ç–æ—Ç –¥–µ–Ω—å –ø–æ–º–æ–≥–∞–µ—Ç –æ–±–Ω–æ–≤–∏—Ç—å –ø—Ä–∏–≤—ã—á–∫–∏, –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –≤–≥–ª—É–±—å —Å–µ–±—è –∏ –ø–æ–Ω—è—Ç—å, –∫—É–¥–∞ –¥–≤–∏–≥–∞—Ç—å—Å—è –¥–∞–ª—å—à–µ.

‚Ä¢ 5 –æ—à–∏–±–æ–∫ –≤ –ø–∏—Ç–∞–Ω–∏–∏ ‚Äî https://app.elsspace.ru/activity/38903
‚Ä¢ –ì–æ—Ä–æ—Å–∫–æ–ø –Ω–∞ –Ω–æ—è–±—Ä—å ‚Äî https://app.elsspace.ru/activity/38683
‚Ä¢ –†–∞–∑–±–æ—Ä –º–∞—Ç—Ä–∏—Ü—ã —Å—É–¥—å–±—ã ‚Äî https://app.elsspace.ru/activity/38883

–°—É–±–±–æ—Ç–∞ ‚Äî –≥–ª—É–±–∏–Ω–∞, –ø–æ–¥–¥–µ—Ä–∂–∫–∞ –∏ –ª–∏—á–Ω—ã–µ –∏–Ω—Å–∞–π—Ç—ã

–í —ç—Ç–æ—Ç –¥–µ–Ω—å —Ç—ã –º–æ–∂–µ—à—å –æ–±—Ä–∞—Ç–∏—Ç—å—Å—è –∫ –æ–ø—ã—Ç—É —ç–∫—Å–ø–µ—Ä—Ç–æ–≤ –∏ –ø–æ–ª—É—á–∏—Ç—å –∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω—É—é –ø–æ–º–æ—â—å –¥–ª—è –≤–∞–∂–Ω—ã—Ö —Ä–µ—à–µ–Ω–∏–π.

‚Ä¢ –ò–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω–∞—è –∫–æ–Ω—Å—É–ª—å—Ç–∞—Ü–∏—è ‚Äî https://app.elsspace.ru/activity/39083
‚Ä¢ –ö–æ–Ω—Å—É–ª—å—Ç–∞—Ü–∏—è –†–æ–¥–æ–ª–æ–≥–∞ ‚Äî https://app.elsspace.ru/activity/38978
‚Ä¢ –≠–Ω–µ—Ä–≥–æ–∑–∞—Å—Ç–∞–≤–∫–∞ –Ω–æ—è–±—Ä—å ‚Äî https://app.elsspace.ru/activity/39008

–í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ ‚Äî –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–µ –∏ —Ä–æ—Å—Ç —á–µ—Ä–µ–∑ —Ä–µ—Å—É—Ä—Å
–¢—ã –º–æ–∂–µ—à—å –ø–æ–∑–∞–±–æ—Ç–∏—Ç—å—Å—è –æ —Ç–µ–ª–µ, –æ–±–Ω–æ–≤–∏—Ç—å —ç–Ω–µ—Ä–≥–∏—é –∏ –Ω–∞–º–µ—Ç–∏—Ç—å –ø–ª–∞–Ω—ã –Ω–∞ –º–µ—Å—è—Ü.

‚Ä¢ –û–±—É—á–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–º—É –º–∞—Å—Å–∞–∂—É ‚Äî https://app.elsspace.ru/activity/39108
‚Ä¢ –î–µ–Ω—å–≥–∏ –ö–∞–∂–¥—ã–π –î–µ–Ω—å ‚Äî –∏–Ω—Ç–µ–Ω—Å–∏–≤ ‚Äî https://app.elsspace.ru/activity/39143
‚Ä¢ –ú–ê–ö-–ø—Ä–∞–∫—Ç–∏–∫—É–º ¬´–¶–µ–ª–∏ —á–µ—Ä–µ–∑ —Ä–µ—Å—É—Ä—Å—ã¬ª ‚Äî https://app.elsspace.ru/activity/33228

–í—ã–±–∏—Ä–∞–π —Ç–æ, —á—Ç–æ –∑–æ–≤—ë—Ç ‚Äî –∏ –Ω–∞—á–∏–Ω–∞–π –Ω–µ–¥–µ–ª—é —Å –∑–∞–±–æ—Ç—ã –æ —Å–µ–±–µ üéØ

–ü–µ—Ä–µ–π–¥–∏ –≤ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ SPACE, –ø–æ—Å–º–æ—Ç—Ä–∏ —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–µ –∏ –¥–æ–±–∞–≤—å –≤ –∫–∞–ª–µ–Ω–¥–∞—Ä—å —Ç–æ, —á—Ç–æ –ø–æ–º–æ–∂–µ—Ç —Ç–µ–±–µ –ø—Ä–æ–∂–∏—Ç—å –Ω–µ–¥–µ–ª—é –æ—Å–æ–∑–Ω–∞–Ω–Ω–æ üëâ
https://app.elsspace.ru/activities

"""

final_text = render_post_with_data(
    template_text,
    events_summary,  # df –∞–≥—Ä–µ–≥–∞—Ç–æ–≤ ["–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å","–ó–Ω–∞—á–µ–Ω–∏–µ"]
    event_list,      # —Å–ø–∏—Å–æ–∫ —Å –¥–∞—Ç–∞–º–∏ –∏ –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏
    events_all,      # —á–∏—Å–ª–æ –∏–ª–∏ df —Å 1 —á–∏—Å–ª–æ–º
    result_df,       # –ù–û–í–û–ï: –±–ª–æ–∫ "3 —Å–ª—É—á–∞–π–Ω—ã—Ö –Ω–∞ –¥–µ–Ω—å" (weekday + ¬´name¬ª + link)
    date_fmt="%d.%m"
)

print(final_text)

# –ü—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏: —Å–º–æ—Ç—Ä–∏, —á—Ç–æ –ø–æ–ø–∞–ª–æ –≤ result_df
# display(result_df)
```
